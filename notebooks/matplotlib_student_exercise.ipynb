## Exercise 5: Subplots (Multi-Plots)

### 🧠 Concept Refresher: Understanding Figure vs. Axes
Before creating subplots, it's essential to understand Matplotlib's Object-Oriented structure:
- **Figure (`fig`)**: The overarching container or "canvas" representing the window or output page. Think of it as a picture frame.
- **Axes (`ax`/`axes`)**: The actual coordinate system (with x and y axes, ticks, titles, labels, and curves) placed on the canvas. A single figure frame can hold one or more axes (plots).

```
 ┌────────────────────────────────────────────────────────┐
 │ Figure (Canvas Frame)                                  │
 │  ┌─────────────────────────┐  ┌─────────────────────┐  │
 │  │ Axes (Left Subplot)     │  │ Axes (Right Subplot)│  │
 │  │  - x/y scales           │  │  - x/y scales       │  │
 │  │  - title, labels        │  │  - title, labels    │  │
 │  └─────────────────────────┘  └─────────────────────┘  │
 └────────────────────────────────────────────────────────┘
```

- **`fig, axes = plt.subplots(nrows, ncols)`**: Creates a canvas partition.
  - The `axes` returned is a NumPy array containing each individual plot object.
  - For a 1D grid (e.g. `1, 2`): `axes[0]` is the first subplot, `axes[1]` is the second.
  - For a 2D grid (e.g. `2, 2`): Access them using double coordinates: `axes[0, 0]` (top-left), `axes[0, 1]` (top-right), etc.
- **`plt.tight_layout()`**: Automatically arranges padding between subplots to prevent labels from overlapping adjacent charts.

**Task:** Create a `1x2` grid of subplots (1 row, 2 columns).
- In the **first subplot** (left), plot a sine wave: `y = np.sin(x)` (color it orange).
- In the **second subplot** (right), plot a cosine wave: `y = np.cos(x)` (color it blue).
- Use `plt.tight_layout()` to make sure the layout looks clean.

````python
# Hint: x = np.linspace(0, 10, 100), use fig, axes = plt.subplots(1, 2)
x = np.linspace(0, 10, 100)

````

<details>
<summary>💡 Click to see solution</summary>

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Plot left axis (index 0)
axes[0].plot(x, np.sin(x), color="orange", label="Sine")
axes[0].set_title("Sine Wave")
axes[0].legend()

# Plot right axis (index 1)
axes[1].plot(x, np.cos(x), color="blue", label="Cosine")
axes[1].set_title("Cosine Wave")
axes[1].legend()

plt.tight_layout()
plt.show()
```
