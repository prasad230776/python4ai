```python
# Tasks 1 & 2
df_sample = df.sample(n=1000, random_state=42)
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df_sample, x="study_hours", y="exam_score", 
                hue="placement_status", size="sleep_hours", 
                sizes=(20, 200), alpha=0.7, palette="Set1")
plt.title("Multivariate: Study vs Exam (Color: Placement, Size: Sleep Hours)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Task 3
plt.figure(figsize=(8, 6))
numeric_cols = df.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Student Metrics")
plt.show()
```
