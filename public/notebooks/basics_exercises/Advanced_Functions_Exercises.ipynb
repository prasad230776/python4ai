#### Exercise 1: Higher-Order Operations using Lambdas

**Task:** 
1. Define a higher-order function `apply_operation(val, op)` that takes a value and a function `op`, and returns `op(val)`.
2. Call `apply_operation(6, lambda x: x ** 2)` to square the number and print it.

````python
# TODO: Write higher-order function and test with lambda
````

<details>
<summary>💡 Click to see solution</summary>

```python
def apply_operation(val, op):
    return op(val)

print(apply_operation(6, lambda x: x ** 2)) # 36
```

#### Exercise 2: Sorting Items by Price

**Task:** Given a list of items and their prices `items = [("Apple", 50), ("Banana", 20), ("Orange", 30)]`:
1. Sort the list by the second element of the tuples (the prices) in ascending order using `sorted()` and a lambda function.
2. Print the sorted list.

````python
# TODO: Sort list using lambda key
````

<details>
<summary>💡 Click to see solution</summary>

```python
items = [("Apple", 50), ("Banana", 20), ("Orange", 30)]
sorted_items = sorted(items, key=lambda item: item[1])
print(sorted_items) # [('Banana', 20), ('Orange', 30), ('Apple', 50)]
```
</details>

#### Exercise 3: Greeting Closures

**Task:** 
1. Write a closure builder function `make_greeter(greeting)` that returns a nested function `greet(name)`.
2. The inner function should return `"{greeting}, {name}!"`.
3. Create a specialized greeter `say_hello = make_greeter("Hello")` and call it with `"Aarav"`.

````python
# TODO: Define and test greeting closure builder
````

<details>
<summary>💡 Click to see solution</summary>

```python
def make_greeter(greeting):
    def greet(name):
        return f"{greeting}, {name}!"
    return greet

say_hello = make_greeter("Hello")
print(say_hello("Aarav")) # Hello, Aarav!
```
</details>

#### Exercise 4: HTML Bold Custom Decorator

**Task:** 
1. Define a decorator named `bold_decorator(func)` that wraps a function returning text.
2. The nested wrapper should enclose the text returned by the original function in HTML bold tags: `f"<b>{original_text}</b>"`.
3. Decorate a function `get_greeting()` that returns `"Hello World"` and test it.

````python
# TODO: Write bold decorator and apply to get_greeting function
````

<details>
<summary>💡 Click to see solution</summary>

```python
def bold_decorator(func):
    def wrapper():
        original_text = func()
        return f"<b>{original_text}</b>"
    return wrapper

@bold_decorator
def get_greeting():
    return "Hello World"

print(get_greeting()) # <b>Hello World</b>
```
</details>
