### Functions as First-Class Objects

#### Example: Functions behaving as Objects
Assigning functions to variables, storing them in lists, passing as arguments, and accessing their special attributes.

```python
def greet(name):
    return f"Hello, {name}!"

# 1. Assign to another variable
say_hello = greet
print(say_hello("Aarav"))

# 2. Store in a list and loop through them
def shout(name):
    return f"HELLO, {name}!"

actions = [greet, shout]
for action in actions:
    print(action("Amit"))

# 3. Pass as an argument (Higher-Order Function)
def run_action(name, action_func):
    return action_func(name)

print(run_action("Dev", greet))

# 4. Access attributes and memory identity
print("Function name attribute:", greet.__name__)
print("Function memory identity:", id(greet))
```

### Lambda Functions

#### Example: Defining Lambda Functions
Comparing standard function structure with one-line anonymous lambdas.

```python
def add(x, y):
    return x + y

add_lambda = lambda x, y: x + y
print(add_lambda(3, 7))
```

#### Example: Real-World Inline Lambda
Applying single-expression mappings to compute taxes/GST on pricing arrays.

```python
prices = [100, 250, 500]
prices_with_gst = list(map(lambda price: price * 1.18, prices))
print(prices_with_gst)
```

#### Example: Multi-line Lambda Expression and Normal Alternative
Writing nested ternaries formatted on multiple lines inside parentheses `()`, and using a standard `def` block for multi-statement functions.

**⚠️ Note on Bad Practice:** Although formatting single expressions across multiple lines using parentheses is syntactically valid, writing complex nested conditionals or multi-line expressions inside a lambda is considered **bad practice** as it hurts code readability and makes debugging difficult. Always prefer standard `def` functions instead.

```python
# 1. Multi-line Lambda Expression (Nested Conditionals - Bad Practice)
classify_score = lambda score: (
    "Excellent" if score >= 90 else
    "Good" if score >= 75 else
    "Pass" if score >= 50 else
    "Fail"
)
print(classify_score(82))
print(classify_score(45))

# 2. Normal Function Alternative for Multiple Statements (Recommended)
def process_user_data(user):
    name = user.get("name", "Guest")
    cleaned_name = name.strip().title()
    print(f"Processing database record for: {cleaned_name}")
    return cleaned_name

user_record = {"name": "  aarav sharma  "}
print(process_user_data(user_record))
```

#### Example: Lambda with Sorted Keys
Using lambdas dynamically to sort tuples by coordinate indices.

```python
points = [(1, 2), (3, 1), (5, 0)]
sorted_points = sorted(points, key=lambda p: p[1])
print(sorted_points)
```

### Closures

#### Example: Multiplier Closures Builder
Declaring nested functions that retain references to enclosing parameters.

```python
def make_multiplier(factor):
    def multiplier(number):
        return number * factor
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(10))
print(triple(10))
```

### Decorators

#### Example: Custom Decorator
Wrapping functions to extend behavior without modifying source code.

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

#### Example: Decorating Functions with Arguments
Passing custom arguments through wrappers using `*args` and `**kwargs`.

```python
def log_arguments(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} returned: {result}")
        return result
    return wrapper

@log_arguments
def add_numbers(a, b):
    return a + b

add_numbers(10, 20)
```

### Variable-Length Arguments

#### Example: print_everything
Collecting extra arguments into tuples and dictionaries dynamically.

```python
def print_everything(*args, **kwargs):
    print("Positional args:", args)
    print("Keyword args:", kwargs)

print_everything(1, 2, 3, name="Amit", age=25)
```
