## Exercise 1: Custom Abstract Validator

**Task:** Create an abstract class `Validator(ABC)` with an abstract method `validate(self, value) -> bool`.
Then, implement a concrete subclass `IntegerValidator(Validator)` that checks if the input value is of type `int`. If true return `True`, otherwise return `False`.

````python
# TODO: Implement Validator interface and IntegerValidator subclass
````

<details>
<summary>💡 Click to see solution</summary>

```python
from abc import ABC, abstractmethod

class Validator(ABC):
    @abstractmethod
    def validate(self, value) -> bool:
        pass

class IntegerValidator(Validator):
    def validate(self, value) -> bool:
        return isinstance(value, int)

val = IntegerValidator()
print(val.validate(10))    # True
print(val.validate("10"))  # False
```

## Exercise 2: Metaclass Attribute Lowercase Inspector

**Task:** Create a custom metaclass `LowercaseAttributeMeta(type)` that validates all custom attributes defined on the class body (ignoring keys starting with `__`). If any attribute key contains uppercase letters, raise a `ValueError`.

````python
# TODO: Implement LowercaseAttributeMeta metaclass
````

<details>
<summary>💡 Click to see solution</summary>

```python
class LowercaseAttributeMeta(type):
    def __new__(mcs, name, bases, namespace):
        for key in namespace:
            if not key.startswith("__"):
                if not key.islower():
                    raise ValueError(f"Attribute '{key}' must be completely lowercase!")
        return super().__new__(mcs, name, bases, namespace)

# Correct usage:
class SafeClass(metaclass=LowercaseAttributeMeta):
    valid_attribute = True
```
