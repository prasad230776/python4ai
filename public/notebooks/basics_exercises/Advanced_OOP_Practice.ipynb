### Concept 1: Abstract Base Classes
Define a PaymentGateway abstract class enforcing `process_payment(amount)` is implemented in child classes.

```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def process_payment(self, amount: float) -> bool:
        pass

class StripePayment(PaymentGateway):
    def process_payment(self, amount: float) -> bool:
        print(f"Processing Rs. {amount} via Stripe")
        return True

payment = StripePayment()
payment.process_payment(1500.0)
```

### Concept 2: Classes as Objects and Dynamic Creation
Creating a class dynamically using `type(name, bases, dict)`.

```python
DynamicCar = type("DynamicCar", (object,),
    {
        "wheels": 4,
        "describe": lambda self: f"I have {self.wheels} wheels."
    }
)

car = DynamicCar()
print(type(DynamicCar))
print(car.describe())
```

### Concept 3: Inspecting `__dict__` and `__annotations__`
Accessing local namespaces and metadata attributes dynamically.

```python
class User:
    username: str
    age: int
    
    def __init__(self, username, age):
        self.username = username
        self.age = age

u = User("Aarav", 24)
print("Instance namespace dict:", u.__dict__)
print("Class type annotations:", User.__annotations__)
```

### Concept 4: Hooking `__new__` and `__call__`
Controlling object allocation and making instances callable.

```python
class Greeter:
    def __new__(cls, *args, **kwargs):
        print("__new__ is allocating memory")
        return super().__new__(cls)
        
    def __init__(self, greeting):
        self.greeting = greeting
        
    def __call__(self, name):
        return f"{self.greeting}, {name}!"

greet = Greeter("Namaste")
print(greet("Aarav"))
```
