#### Exercise 1: Trigonometric Cosine of 60 Degrees

**Task:** 
1. Import only `radians` and `cos` from the `math` module.
2. Convert 60 degrees to radians using the `radians()` function.
3. Compute the cosine of the resulting value and print it.

````python
# TODO: Import and calculate cosine of 60 degrees
````

<details>
<summary>💡 Click to see solution</summary>

```python
from math import radians, cos

rad = radians(60)
print(cos(rad)) # 0.5000000000000001 (approx 0.5)
```

#### Exercise 2: Random Team Selection

**Task:** Given a list `members = ["Aarav", "Amit", "Dev", "Priya"]`:
1. Use `random.choice()` to randomly pick a single team leader.
2. Use `random.sample()` to pick 2 unique members as helpers.
3. Print both outputs.

````python
# TODO: Select random leader and sample helpers
````

<details>
<summary>💡 Click to see solution</summary>

```python
import random

members = ["Aarav", "Amit", "Dev", "Priya"]
leader = random.choice(members)
helpers = random.sample(members, 2)

print("Leader:", leader)
print("Helpers:", helpers)
```
</details>

#### Exercise 3: Days to End of Year

**Task:** 
1. Create a `datetime.date` object representing the current date.
2. Create another `datetime.date` object representing December 31st of the current year (2026).
3. Subtract the current date from the end-of-year date and print how many days are left.

````python
# TODO: Calculate remaining days until end of year
````

<details>
<summary>💡 Click to see solution</summary>

```python
import datetime

today = datetime.date.today()
end_of_year = datetime.date(today.year, 12, 31)
days_left = (end_of_year - today).days
print("Days left in year:", days_left)
```
</details>

#### Exercise 4: Directory Builder and Path Join

**Task:** 
1. Check if a directory named `"log_output"` exists. If not, create it.
2. Generate a joined directory path pointing to `"output.log"` inside the `"log_output"` folder using `os.path.join()`.
3. Print the joined path.

````python
# TODO: Check/create directory and construct log path
````

<details>
<summary>💡 Click to see solution</summary>

```python
import os

folder = "log_output"
if not os.path.exists(folder):
    os.mkdir(folder)
    
log_path = os.path.join(folder, "output.log")
print("Constructed log path:", log_path)
```
