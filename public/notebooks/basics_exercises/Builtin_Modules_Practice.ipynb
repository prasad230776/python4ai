### Import Patterns

#### Example: Entire Module Import
Importing a whole module and accessing its tools using dot notation.

```python
import math
print(math.sqrt(16))
```

#### Example: Specific Items Import
Importing specific functions/constants directly into the namespace.

```python
from math import sqrt, pi
print(sqrt(16))
print(pi)
```

### The math Module

#### Example: Constants
Accessing standard constants defined on the math module.

```python
import math
print(math.pi)
print(math.e)
```

#### Example: Rounding Functions
Rounding UP and DOWN to nearest integers.

```python
import math
print(math.ceil(4.2))
print(math.floor(4.8))
```

#### Example: Power and Roots
Computing square roots and exponential powers.

```python
import math
print(math.sqrt(16))
print(math.pow(2, 3))
```

### The random Module

#### Example: Generating Random Numbers
Generating random floats and integers across bounds.

```python
import random
print(random.random())
print(random.randint(1, 10))
print(random.uniform(1.5, 5.5))
```

#### Example: Selecting from a List
Randomly choosing single and multiple unique items.

```python
import random
fruits = ["apple", "banana", "cherry", "date"]
print(random.choice(fruits))
print(random.sample(fruits, 2))
```

#### Example: Shuffling a List
Reordering elements in-place dynamically.

```python
import random
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
```

### The datetime Module

#### Example: Current Dates and Specific Instantiations
Retrieving timestamps and declaring specific date components.

```python
import datetime
today = datetime.date.today()
print("Today:", today)
now = datetime.datetime.now()
print("Now:", now)

birthday = datetime.date(1995, 10, 25)
print("Birthday year:", birthday.year)
```

#### Example: Date Arithmetic
Evaluating date deltas and offset modifications using `timedelta`.

```python
import datetime
today = datetime.date.today()
birthday = datetime.date(1995, 10, 25)

ten_days_later = today + datetime.timedelta(days=10)
print("Ten days later:", ten_days_later)
delta = today - birthday
print("Days elapsed since birthday:", delta.days)
```

#### Example: Formatting Dates as Strings
Formatting datetime objects to string representations using `.strftime()`.

```python
import datetime
now = datetime.datetime.now()
formatted = now.strftime("%d-%m-%Y %H:%M")
print("Formatted string:", formatted)
```

#### Example: Parsing Strings to Dates
Reconstructing datetime objects from string patterns using `.strptime()`.

```python
import datetime
date_string = "2026-07-19"
parsed_date = datetime.datetime.strptime(date_string, "%-Y-%m-%d")
print("Parsed datetime object:", parsed_date)
```

### The os Module

#### Example: Directory Management
Getting directories, listing child assets, and creating local folders.

```python
import os
print("Current working directory:", os.getcwd())
print("List directory files:", os.listdir(".")[:5])

if not os.path.exists("temp_folder"):
    os.mkdir("temp_folder")
```

#### Example: Path Modifications
Safely joining directory scopes and dissecting folder structures.

```python
import os
full_path = os.path.join("temp_folder", "data.txt")
print("Joined path:", full_path)

dir_name = os.path.dirname(full_path)
file_name = os.path.basename(full_path)
print(f"Dir: {dir_name}, File: {file_name}")
```
