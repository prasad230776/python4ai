#### Exercise 1: Student Class Blueprint

**Task:** Create a class named `Student` containing:

#### Exercise 2: Class Attributes and Helper Methods

**Task:** Add the following elements to the `Student` class:
- A class attribute `school_name` initialized to `"National Public School"`.
- A class method `change_school(cls, new_name)` using the `@classmethod` decorator to modify the school name.
- A static method `is_passing_score(score)` using the `@staticmethod` decorator that returns `True` if `score >= 40` and `False` otherwise.

````python
# TODO: Update Student class with class attribute, class method, and static method
````

<details>
<summary>💡 Click to see solution</summary>

```python
class Student:
    school_name = "National Public School"

    def __init__(self, name, score):
        self.name = name
        self.score = score

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name

    @staticmethod
    def is_passing_score(score):
        return score >= 40

print(Student.school_name) # National Public School
Student.change_school("Delhi Public School")
print(Student.school_name) # Delhi Public School
print(Student.is_passing_score(45)) # True
```
</details>

#### Exercise 4: Subclassing & Single Inheritance

**Task:** 
1. Create a parent class `Person` with `__init__(self, name)`.
2. Create a child class `Student` inheriting `Person` that takes both `name` and `grade` in its initializer, using `super().__init__(name)` to pass the name up to the parent constructor.
3. Instantiate `Student("Dev", "12th Grade")` and print name and grade.

````python
# TODO: Create Person and Student class using single inheritance
````

<details>
<summary>💡 Click to see solution</summary>

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade

stud = Student("Dev", "12th Grade")
print("Name:", stud.name, "| Grade:", stud.grade)
```
</details>
