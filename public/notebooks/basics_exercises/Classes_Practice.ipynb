### Functional vs Object-Oriented Programming Paradigms

#### Example: Functional Programming Approach
Applying independent transforms to state-free data.

```python
def clean_text(text):
    return text.strip().lower()

def remove_punctuation(text):
    return text.replace(".", "").replace(",", "")

result = remove_punctuation(clean_text("  Hello, World.  "))
print("Functional result:", result)
```

#### Example: Object-Oriented Programming Approach
Bundling state and cleaning methods inside a single class.

```python
class TextProcessor:
    def __init__(self, text):
        self.text = text
    
    def clean(self):
        self.text = self.text.strip().lower()
        return self
    
    def remove_punctuation(self):
        self.text = self.text.replace(".", "").replace(",", "")
        return self

processor = TextProcessor(text="  Hello, World.  ")
result = processor.clean().remove_punctuation().text
print("OOP result:", result)
```

### Creating Classes and Instances

#### Example: Simple Account Class Definition
Creating a blueprint and instantiating it.

```python
class Account:
    pass

my_account = Account()
print("Object type:", type(my_account))
```

#### Example: Dynamic Attribute Assignment
Adding attributes to an object instance on-the-fly.

```python
my_account = Account()
my_account.owner = "Aarav"
my_account.balance = 5000.0

print("Owner:", my_account.owner)
print("Balance:", my_account.balance)
```

#### Example: Adding Behavior (Instance Methods)
Defining methods that act on instance variables using `self`.

```python
class Account:
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited Rs. {amount}. New Balance: Rs. {self.balance}")

my_account = Account()
my_account.balance = 5000.0
my_account.deposit(500.0)
```

#### Example: The `__init__` Initializer Method
Initializing attributes automatically during object creation.

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

my_account = Account("Aarav", 5000.0)
print(my_account.owner, "has Rs.", my_account.balance)
```

### Attributes & Methods Types

#### Example: Class Attributes vs Instance Attributes
Comparing class-level shared attributes with instance-level private fields.

```python
class Account:
    bank_name = "State Bank of India"  # Class attribute (shared)

    def __init__(self, owner, balance):
        self.owner = owner             # Instance attribute (unique)
        self.balance = balance

acc1 = Account("Aarav", 5000)
acc2 = Account("Amit", 7000)
print(acc1.bank_name, "|", acc2.bank_name)
```

#### Example: Class Methods (`@classmethod`)
Methods bound to the class itself using `cls` to modify class attributes.

```python
class Account:
    bank_name = "State Bank of India"

    @classmethod
    def change_bank(cls, new_bank):
        cls.bank_name = new_bank

Account.change_bank("ICICI Bank")
print(Account.bank_name)
```

#### Example: Static Methods (`@staticmethod`)
Helper functions inside classes that do not require references to `self` or `cls`.

```python
class Account:
    @staticmethod
    def is_valid_transaction(amount):
        return amount > 0

print("Is 500 valid?:", Account.is_valid_transaction(500))
```

#### Example: Data Encapsulation and Name Mangling
Exploring public, protected (`_`), and private (`__`) attribute conventions and name mangling lookup.

```python
class Account:
    def __init__(self, owner, pin):
        self.owner = owner      # Public
        self._balance = 5000.0  # Protected
        self.__pin = pin        # Private (Name Mangled)

acc = Account("Aarav", 1234)
print("Public owner:", acc.owner)
print("Protected balance:", acc._balance)
# print(acc.__pin) # Raises AttributeError!

# Access via mangled name:
print("Mangled PIN access:", acc._Account__pin)
```

### Inheritance Types & MRO

#### Example: Basic Animal and Dog Inheritance
 A child class inheriting basic methods from a parent.

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def eat(self):
        return f"{self.name} is eating"
    
    def sleep(self):
        return f"{self.name} is sleeping"

class Dog(Animal):
    def bark(self):
        return f"{self.name} says woof!"

my_dog = Dog("Buddy")
print(my_dog.eat())
print(my_dog.sleep())
print(my_dog.bark())
```

#### Example: Adding Attributes with super().__init__()
Calling parent initializers to set up base fields cleanly.

```python
class Animal:
    def __init__(self, name):
        self.name = name
        self.is_pet = True

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Pass name to parent
        self.breed = breed
    
    def describe(self):
        return f"{self.name} is a {self.breed}"

golden = Dog("Buddy", "Golden Retriever")
print(golden.describe())
print("Is Pet?:", golden.is_pet)
```

#### Example: Overriding Methods
Customizing parent methods in subclasses.

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def make_sound(self):
        return f"{self.name} makes a sound"

class Dog(Animal):
    def make_sound(self):
        return f"{self.name} barks: Woof!"

class Cat(Animal):
    def make_sound(self):
        return f"{self.name} meows: Meow!"

generic = Animal("Something")
dog = Dog("Buddy")
cat = Cat("Whiskers")
print(generic.make_sound())
print(dog.make_sound())
print(cat.make_sound())
```

#### Example: Practical AI Model Classes
Organizing text processing base models.

```python
class BaseModel:
    def __init__(self, model_name):
        self.model_name = model_name
        self.is_loaded = False
    
    def load(self):
        print(f"Loading {self.model_name}...")
        self.is_loaded = True

class TextModel(BaseModel):
    def __init__(self, model_name, max_length=100):
        super().__init__(model_name)
        self.max_length = max_length
    
    def process_text(self, text):
        if not self.is_loaded:
            self.load()
        if len(text) > self.max_length:
            text = text[:self.max_length]
        return f"Processed: {text}"

model = TextModel(model_name="gpt-3.5-turbo")
print(model.process_text("Hello world"))
```

#### Example: Multiple Inheritance Conflict Resolution (MRO)
Using C3 Linearization MRO search order with `ParentA` and `ParentB` structures.

```python
class ParentA:
    def greet(self):
        return "Hello from Parent A"

class ParentB:
    def greet(self):
        return "Hello from Parent B"

class Child(ParentA, ParentB):
    pass

c = Child()
print("Greeting:", c.greet())
print("MRO Resolution Order:", Child.mro())
```
