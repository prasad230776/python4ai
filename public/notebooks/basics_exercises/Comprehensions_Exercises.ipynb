#### Exercise 1: Filter Uppercase Names

**Task:** Given a list of strings `names = ["Aarav", "banana", "Dev", "grape"]`:
1. Write a list comprehension that extracts only the words that start with an uppercase letter (hint: check `word[0].isupper()`).
2. Print the final list.

````python
# TODO: Extract capitalized words using list comprehension
````

<details>
<summary>💡 Click to see solution</summary>

```python
names = ["Aarav", "banana", "Dev", "grape"]
capitalized = [name for name in names if name[0].isupper()]
print(capitalized) # ['Aarav', 'Dev']
```
</details>

#### Exercise 2: Name Length Mapping

**Task:** Given a list of strings `words = ["Aarav", "Dev", "Amit"]`:
1. Write a dictionary comprehension that maps each string to its length (e.g., `{"Dev": 3}`).
2. Print the dictionary.

````python
# TODO: Map words to their length using dict comprehension
````

<details>
<summary>💡 Click to see solution</summary>

```python
words = ["Aarav", "Dev", "Amit"]
word_lengths = {word: len(word) for word in words}
print(word_lengths) # {'Aarav': 5, 'Dev': 3, 'Amit': 4}
```
</details>

#### Exercise 3: Custom Even Numbers Iterator

**Task:** Write a custom iterator class `EvenCounter` that takes a `limit`:
- It should yield even numbers starting from `2` up to `limit`.
- Once it exceeds `limit`, raise `StopIteration`.

````python
# TODO: Define and test the EvenCounter iterator
````

<details>
<summary>💡 Click to see solution</summary>

```python
class EvenCounter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.current += 2
        if self.current <= self.limit:
            return self.current
        else:
            raise StopIteration

for num in EvenCounter(8):
    print(num) # 2, 4, 6, 8
```
</details>

#### Exercise 4: Custom Managed File Context Manager

**Task:** Create a custom context manager `managed_file(filename, mode)` using `@contextmanager`:
- It should open the file with the specified filename and mode.
- Yield the file resource to the `with` block.
- Ensure that the file is always closed in a `finally` block.

````python
# TODO: Write managed_file context manager and test it
````

<details>
<summary>💡 Click to see solution</summary>

```python
from contextlib import contextmanager

@contextmanager
def managed_file(filename, mode):
    f = open(filename, mode)
    try:
        yield f
    finally:
        f.close()
        print("File closed safely.")

with managed_file("exercise_temp.txt", "w") as f:
    f.write("Hello from context manager!")
```
</details>
