### List Comprehensions

#### Example: Squares and Even Squares
Generating list values concisely using loops and conditional filters.

```python
# Traditional loop
squares_loop = []
for x in range(5):
    squares_loop.append(x * x)
print("Loop squares:", squares_loop)
```

```python
# Pythonic List Comprehension
squares = [x * x for x in range(5)]
print("Comprehension squares:", squares)
```

```python
# With conditionals (only even squares)
even_squares = [x * x for x in range(10) if x % 2 == 0]
print("Even squares:", even_squares)
```

### Dictionary Comprehensions

#### Example: Mapping Squares
Dynamically building dictionary collections using key-value pair comprehensions.

```python
squares_dict = {x: x * x for x in range(5)}
print(squares_dict)
```

### Set Comprehensions

#### Example: Unique Name Lengths
Filtering unique lengths using set comprehensions.

```python
names = ["alice", "bob", "alice", "charlie"]
unique_lengths = {len(name) for name in names}
print(unique_lengths)
```

### Iterators

#### Example: Custom Counter Iterator
Implementing the iterator protocol using `__iter__` and `__next__`.

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            self.current += 1
            return self.current
        else:
            raise StopIteration

for num in Counter(3):
    print(num)
```

### Generators

#### Example: Generator Functions
Pausing execution contexts with `yield` statements.

```python
def simple_generator():
    yield "First"
    yield "Second"
    yield "Third"
```

```python
gen = simple_generator()
print(next(gen))
print(next(gen))
```

#### Example: Generator Expressions
Using generator syntax inside `()` to yield items lazily.

```python
# Generator expression (computes values lazily on the fly)
gen_exp = (x * x for x in range(1000000))

print(next(gen_exp))
print(next(gen_exp))
```

### Context Managers

#### Example: File handling with context managers
Ensuring resources are closed cleanly using `with` blocks.

```python
with open("data_temp.txt", "w") as file:
    file.write("Pythonic Programming")
print("File write complete")
```

#### Example: Custom Context Manager
Creating resource wrappers using `@contextmanager`.

```python
from contextlib import contextmanager

@contextmanager
def managed_resource():
    print("Acquiring resource... 🔑")
    try:
        yield "Active Resource"
    finally:
        print("Releasing resource... 🔒")

with managed_resource() as res:
    print(f"Using: {res}")
```
