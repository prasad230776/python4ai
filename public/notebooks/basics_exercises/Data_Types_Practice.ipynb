## 1. Numbers (Integers & Floats)

### Concept 1: Integers vs. Floats & Basic Operations
Identifying types and regular vs floor division.

```python
age = 25
price = 19.99
print("Type of age:", type(age))
print("Type of price:", type(price))

# Regular vs Floor Division
print("10 / 3 =", 10 / 3)
print("10 // 3 =", 10 // 3)
```

### Concept 2: Numbers are Objects
Even integers and floats are objects and have built-in methods.

```python
# Checking bit length of integer
print("Bit length of 42:", (42).bit_length())

# Checking if floats are whole numbers
print("Is 4.0 integer?", (4.0).is_integer())
print("Is 4.5 integer?", (4.5).is_integer())
```

### Concept 3: Integer Caching (Object Sharing)
Python shares memory locations (caches) for small integers between -5 and 256.

```python
# Small integers (share memory)
a = 10
b = 10
print("id(a) == id(b):", id(a) == id(b))
print("a is b:", a is b)

# Larger integers (do not share memory)
x = 500
y = 500
print("x is y:", x is y)
```

### Concept 4: Floating-Point Precision Issues
Decimals that are repeating binary fractions (like 0.1) cause small rounding errors, while exact powers of 2 (like 0.25 and 0.5) do not.

```python
# Repeating binary fraction issue
print("0.1 + 0.2 =", 0.1 + 0.2)
print("0.1 + 0.2 == 0.3?", 0.1 + 0.2 == 0.3)

# Exact binary representation
print("0.25 + 0.5 =", 0.25 + 0.5)
print("0.25 + 0.5 == 0.75?", 0.25 + 0.5 == 0.75)
```

## 2. Strings (Text)

```python
# Single, Double, and Triple quotes
simple_str = "Hello"
apostrophe_str = "It's Python!"
multiline_str = """This is a
multi-line string"""

# Combining and repeating strings
first_name = "Rahul"
last_name = "Sharma"
full_name = first_name + " " + last_name
stars = "*" * 10

# String length and conversion
text_len = len(full_name)
age_str = "I am " + str(25) + " years old."

print(full_name)
print(stars)
print("Length of name:", text_len)
print(age_str)
```

## 3. Booleans (True / False)

```python
# Assigning Booleans
is_active = True
is_admin = False
print("is_active type:", type(is_active))

# Comparisons returning Booleans
age = 20
can_vote = age >= 18
is_exact_age = age == 20

print("Can vote?", can_vote)
print("Is exactly 20?", is_exact_age)
```
