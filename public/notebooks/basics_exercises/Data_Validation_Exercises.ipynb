#### Exercise 2: Book Dataclass Definition

**Task:** 
1. Define a dataclass `Book` with attributes: `title: str`, `author: str`, `pages: int`, and `price: float = 0.0` (with default value `0.0`).
2. Instantiate a `Book` and print the object to verify the generated `__repr__` output formatting.

````python
# TODO: Implement Book dataclass and instantiate it
````

<details>
<summary>💡 Click to see solution</summary>

```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    pages: int
    price: float = 0.0

book = Book("Python Basics", "Aarav Sharma", 300, 450.00)
print(book)
```
</details>

#### Exercise 4: Dataclasses with Collections

**Task:** 
1. Define a dataclass `Course` with attributes: `title: str`, `students: list[str]`, and `topics: set[str]`.
2. Create an instance of `Course` and print its details.

````python
# TODO: Define Course dataclass and print instance details
````

<details>
<summary>💡 Click to see solution</summary>

```python
from dataclasses import dataclass

@dataclass
class Course:
    title: str
    students: list[str]
    topics: set[str]

course = Course(
    title="Python Internals",
    students=["Aarav", "Amit"],
    topics={"Memory Model", "Garbage Collection"}
)
print(course)
```
</details>
