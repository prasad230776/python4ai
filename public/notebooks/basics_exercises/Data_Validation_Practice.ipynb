### Basic Type Hinting

#### Example: Variable Annotations
Annotating simple variables with types.

```python
age: int = 25
name: str = "Amit"
is_active: bool = True
```

#### Example: Function Annotations
Annotating parameters and return types.

```python
def calculate_total(price: float, quantity: int) -> float:
    return price * quantity

print(calculate_total(19.99, 3))
```

### Dataclasses

#### Example: Traditional Class vs. Dataclass
Comparing boilerplate traditional classes to cleaner `@dataclass` syntax.

```python
from dataclasses import dataclass

# Clean Dataclass Equivalent
@dataclass
class Product:
    name: str
    price: float
    stock: int

item = Product("Laptop", 999.99, 5)
print(item)  # Auto-generated formatted __repr__!

item_two = Product("Laptop", 999.99, 5)
print("Are items equal?:", item == item_two)  # Auto-generated values comparator __eq__!
```

#### Example: Default Values in Dataclasses
Defining fallback values for fields.

```python
from dataclasses import dataclass

@dataclass
class User:
    username: str
    email: str
    is_admin: bool = False

u = User("aarav123", "aarav@example.com")
print(u)
```

#### Example: Dataclasses with Collections
Annotating attributes with list, set, and dictionary collections.

```python
from dataclasses import dataclass

@dataclass
class StudentProfile:
    name: str
    scores: list[int]
    skills: set[str]
    metadata: dict[str, str]

profile = StudentProfile(
    name="Amit",
    scores=[95, 88, 92],
    skills={"Python", "SQL", "Python"},
    metadata={"cohort": "2026", "advisor": "Dr. Smith"}
)

print(profile.scores)
print(profile.skills)
```
