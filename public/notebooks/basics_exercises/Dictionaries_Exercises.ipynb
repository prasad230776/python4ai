## Exercise 1: Grades Lookup and Defaults

**Task:** Create a dictionary named `grades` with the keys `"Math"`: 90 and `"Science"`: 88. 
1. Access the grade for `"Math"` and print it.
2. Try to access the grade for `"English"` using `.get()`, providing a default fallback score of `40` if it doesn't exist, and print the result.
3. Add a new subject `"English"` with a score of `82` to the dictionary, then print the updated dictionary.

````python
# TODO: Perform dictionary operations
````

<details>
<summary>💡 Click to see solution</summary>

```python
grades = {"Math": 90, "Science": 88}
print("Math score:", grades["Math"]) # 90

english_score = grades.get("English", 40)
print("English score (fallback):", english_score) # 40

grades["English"] = 82
print("Updated grades:", grades) # {'Math': 90, 'Science': 88, 'English': 82}
```

## Exercise 2: Inventory Updating and Views

**Task:** Given a store inventory `stock = {"notebooks": 50, "pens": 100}`:
1. Update the inventory in bulk using `.update()` to change 'notebooks' to 45 and add 'erasers' with a count of 30.
2. Print the list of all keys in the inventory.
3. Pop 'pens' from the inventory, and print both the popped value and the remaining inventory.

````python
# TODO: Perform update, list keys, and pop from stock dictionary
stock = {"notebooks": 50, "pens": 100}
````

<details>
<summary>💡 Click to see solution</summary>

```python
stock = {"notebooks": 50, "pens": 100}

# 1. Update in bulk
stock.update({"notebooks": 45, "erasers": 30})

# 2. Print keys
print("Stock keys:", list(stock.keys())) # ['notebooks', 'pens', 'erasers']

# 3. Pop item
pens_count = stock.pop("pens")
print("Popped pens:", pens_count) # 100
print("Remaining stock:", stock)  # {'notebooks': 45, 'erasers': 30}
```
