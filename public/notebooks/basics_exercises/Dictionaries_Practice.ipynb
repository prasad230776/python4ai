# Python Dictionaries - Follow-Along Practice

Use this notebook to practice using Python dictionaries: key-value storage, various creation patterns, lookups, modifications, and all key methods (`keys()`, `values()`, `items()`, `update()`, `pop()`, `clear()`, `len()`).
Each concept is in a separate cell, followed by a blank code cell for you to practice.

### Concept 1: Various Ways of Creating Dictionaries
Explore literal curly braces, dict() constructors with keywords, dict() with list of tuples, and dict.fromkeys() methods.

```python
# 1. Curly braces
d1 = {"a": 1, "b": 2}

# 2. Constructor keyword args
d2 = dict(a=1, b=2)

# 3. Constructor list of tuples
d3 = dict([("a", 1), ("b", 2)])

# 4. From keys with default values
d4 = dict.fromkeys(["a", "b"], 0)

print(d1, d2, d3, d4)
```

### Concept 2: Accessing Values and the `.get()` Method
Accessing values using square brackets or safely with `.get()` to avoid raising errors.

```python
student = {"name": "Aarav", "score": 85}
print("Name:", student["name"])
print("Score:", student.get("score"))
print("Age (safe lookup fallback):", student.get("age", "Not Available"))
```

### Concept 3: Dictionary Views (`keys()`, `values()`, `items()`)
Extracting keys, values, or key-value tuples from a dictionary.

```python
profile = {"name": "Amit", "city": "Delhi", "job": "Engineer"}
print("Keys View:", profile.keys())
print("Values View:", profile.values())
print("Items View:", profile.items())
```

### Concept 4: Updating Dictionaries and `.update()`
Adding/modifying individual keys or merging multiple key-value pairs at once using `.update()`.

```python
cart = {"apple": 2, "banana": 5}
cart["orange"] = 3                         # Add single key
cart.update({"apple": 4, "mango": 2})      # Update and add multiple keys
print("Updated cart:", cart)
print("Total unique products:", len(cart)) # len() function
```

### Concept 5: Deleting Elements (`pop()`, `del`, `clear()`)
Removing items using `.pop(key)` to retrieve the value, `del` keyword, or emptying the dictionary entirely using `.clear()`.

```python
student = {"name": "Dev", "age": 22, "score": 95, "city": "Chennai"}

popped_val = student.pop("score") # Remove and return value
del student["city"]              # Remove using del keyword
print("After partial removals:", student)
print("Popped score:", popped_val)

student.clear()                  # Empty the dictionary
print("After clear:", student)
```
