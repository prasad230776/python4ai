#### Exercise 1: Safe Division Utility

**Task:** Write a function `safe_divide(a, b)` that performs division `a / b` inside a `try` block:
- If a `ZeroDivisionError` occurs, return `None` and print `"Warning: Division by zero attempted."`. 
- If a `TypeError` occurs, return `None` and print `"Warning: Invalid type input."`.
- Otherwise (no exception), return the division result.

````python
# TODO: Write and test the safe_divide function
````

<details>
<summary>💡 Click to see solution</summary>

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Warning: Division by zero attempted.")
        return None
    except TypeError:
        print("Warning: Invalid type input.")
        return None

print(safe_divide(10, 2))  # 5.0
print(safe_divide(10, 0))  # None
print(safe_divide(10, "2")) # None
```
</details>

def parse_input(text):
    try:
        value = int(text)
    except ValueError:
        print("Failed conversion.")
    else:
        print(f"Success: converted value is {value}")
    finally:
        print("Conversion step executed.")

#### Exercise 3: Valid Age Exception

**Task:** 
1. Define a custom exception `InvalidAgeError` inheriting from `Exception`.
2. Write a function `register_age(age)` that raises `InvalidAgeError` if `age < 0` or `age > 150`.
3. Call `register_age(-5)` inside a try-except block and print the exception's message.

````python
# TODO: Define exception and call function with invalid age
````

<details>
<summary>💡 Click to see solution</summary>

```python
class InvalidAgeError(Exception):
    pass

def register_age(age):
    if age < 0 or age > 150:
        raise InvalidAgeError(f"Age {age} is outside the valid range (0-150)")

try:
    register_age(-5)
except InvalidAgeError as e:
    print("Error occurred:", e)
```
</details>

2. For the list `vals = [10, 20]`, access index `3` safely by using a `try-except IndexError` block.

````python
# TODO: Safely access missing key and index
````

```python
info = {"name": "Aarav"}
vals = [10, 20]

# Part 1: dict check
if "email" in info:
    print(info["email"])
else:
    print("email key is missing")

# Part 2: list index
try:
    print(vals[3])
except IndexError:
    print("Index 3 is out of range")
```
