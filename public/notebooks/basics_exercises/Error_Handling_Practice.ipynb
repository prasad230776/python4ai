### Basic Exception Handling

#### Example: Basic try-except Structure
Catching a general exception using try-except.

```python
try:
    # Attempting to access undefined variable
    print(undefined_variable)
except Exception:
    print("Something went wrong")
```

#### Example: Catching Specific Errors
Catching specific error types like `ValueError`.

```python
try:
    # Simulating value parsing error
    age = int("not-a-number")
except ValueError:
    print("Please enter a number")
```

#### Example: Catching Multiple Error Types
Routing different exceptions to separate except blocks.

```python
try:
    # Can cause FileNotFoundError, ValueError, or ZeroDivisionError
    number = int("0")
    result = 100 / number
    print(f"Result: {result}")
except FileNotFoundError:
    print("Could not find file")
except ValueError:
    print("Not a valid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

#### Example: The `else` Clause
Running code only if no exceptions are raised inside the `try` block.

```python
try:
    data = "Important Data String"
except FileNotFoundError:
    print("File not found")
else:
    print(f"Success! Data length: {len(data)}")
```

#### Example: The `finally` Clause
Executing cleanup logic regardless of whether an error occurred.

```python
try:
    print("Performing operation...")
    result = 1 / 0
except ZeroDivisionError:
    print("Caught division by zero!")
finally:
    print("Cleanup complete")
```

#### Example: Logging and Re-raising Critical Errors
Intercepting errors for debugging and forwarding them upwards using `raise`.

```python
try:
    print("Critical computation...")
    # raise ValueError("Invalid state")
except ValueError as e:
    print(f"Logging critical error: {e}")
    # raise # Let it propagate
```

### Common Errors & Fixes

#### Example: FileNotFoundError
Verifying file checks vs try-except blocks.

```python
import os

# Pre-check
if os.path.exists("missing.txt"):
    with open("missing.txt", "r") as f:
        content = f.read()
else:
    print("File not found (via os.path.exists check)")

# Try-except
try:
    with open("missing.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("File not found (via try-except block)")
```

#### Example: ValueError
Handling values with inappropriate configurations.

```python
try:
    val = int("12.5") # Decimal string direct cast fails
except ValueError:
    # Fix: float conversion first
    val = int(float("12.5"))
    print("Value recovered:", val)
```

#### Example: KeyError
Dictionary key checks and safe access methods.

```python
user = {"name": "Aarav", "age": 25}

# Using get()
email = user.get("email", "no-email@example.com")
print("Email via get():", email)

# Checking existence
if "email" in user:
    print(user["email"])
else:
    print("Email key not found")
```

#### Example: IndexError
Safe indexing checks.

```python
numbers = [1, 2, 3]

# Safe length boundary checking
if len(numbers) > 5:
    print(numbers[5])
else:
    print("Index out of bounds")
```

#### Example: TypeError
Verifying type checks before mismatched actions.

```python
try:
    # Cannot concatenate string and integer
    result = "hello" + 5
except TypeError:
    # Fix by explicit casting
    result = "hello" + str(5)
    print("Recovered Type Result:", result)
```

#### Example: AttributeError
Inspecting method bindings on targets using `hasattr()`.

```python
text = "hello"

# Check if append exists on string (it does not)
if hasattr(text, "append"):
    text.append("!")
else:
    # Correct concatenation
    text = text + "!"
    print("Attribute corrected:", text)
```

### Custom Exceptions

#### Example: Simple Custom Exception
Creating custom classes inheriting from base `Exception`.

```python
class MyCustomError(Exception):
    pass

try:
    raise MyCustomError("Custom failure!")
except MyCustomError as e:
    print("Caught:", e)
```

#### Example: Insufficient Balance Error Scenario
Defining and handling complex errors with custom attributes.

```python
class InsufficientBalanceError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Attempted to withdraw Rs. {amount} but only have Rs. {balance}")

class BankAccount:
    def __init__(self, balance):
        self.balance = balance
        
    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientBalanceError(self.balance, amount)
        self.balance -= amount
        print(f"Withdrew Rs. {amount}. Remaining: Rs. {self.balance}")

account = BankAccount(100.0)
try:
    account.withdraw(150.0)
except InsufficientBalanceError as error:
    print("Transaction Failed:", error)
    print("Shortage:", error.amount - error.balance)
```
