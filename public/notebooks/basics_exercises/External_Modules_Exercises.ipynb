#### Exercise 1: Querying httpbin API

**Task:** 
1. Import the external `requests` library.
2. Perform an HTTP GET request to `https://httpbin.org/get` using `requests.get()`.
3. Print the response status code.
4. Parse the response payload as a dictionary using `.json()` and print the value of the `"url"` key.

````python
# TODO: Query https://httpbin.org/get and print url value
````

<details>
<summary>💡 Click to see solution</summary>

```python
import requests

response = requests.get("https://httpbin.org/get")
print("Status Code:", response.status_code)
data = response.json()
print("Requested URL:", data["url"])
```

#### Exercise 2: Catching Missing Package Exception

**Task:** 
1. Write a try-except block that attempts to import a non-existent package `super_cool_library`.
2. Catch the specific `ModuleNotFoundError` raised by Python.
3. Inside the exception block, print a friendly message: `"Please run: pip install super_cool_library"`.

````python
# TODO: Write try-except block wrapping import statement
````

<details>
<summary>💡 Click to see solution</summary>

```python
try:
    import super_cool_library
except ModuleNotFoundError:
    print("Please run: pip install super_cool_library")
```
</details>
