### Using External Packages in Code

#### Example: Fetching Data with requests
Importing and calling the popular external HTTP library `requests` to query GitHub API events.

```python
import requests

response = requests.get("https://api.github.com/events")
print("Status Code:", response.status_code)
print("Data type of response content:", type(response.json()))
print("First event keys:", list(response.json()[0].keys()) if response.status_code == 200 else "Failed to load")
```

### Handling Missing Packages Safely

#### Example: ModuleNotFoundError Try-Except Guard
Using try-except validation wrappers to detect absent packages and instruct users how to install them.

```python
try:
    import non_existent_package
except ModuleNotFoundError:
    print("Module 'non_existent_package' is not installed in the current environment.")
    print("To install it, run: pip install non_existent_package")
```

### Inspecting Package Locations

#### Example: Checking Module filepath (`__file__`)
Inspecting filepaths of imported packages to make sure they are loaded from virtual environments and not local files (shadowing).

```python
import requests
print("Module filepath:", requests.__file__)
```
