# Practice with Python f-Strings

This notebook will help you explore additional formatting options using f-strings in Python.
Each section introduces a new feature you can try!

## 1. Binary, Octal, and Hexadecimal
You can display numbers in different bases using `b`, `o`, or `x`:

```python
num = 42
print(f"Binary: {num:b}")
print(f"Octal: {num:o}")
print(f"Hexadecimal: {num:x}")
```

## 2. Padding Numbers with Zeros
Use `0` before width to pad numbers with zeros:

```python
score = 7
print(f"Score: {score:03}")
```

## 3. Sign Formatting
Use `+` to always show sign:

```python
pos_balance = 42
neg_balance = -21
print(f"Normal: {pos_balance} | {neg_balance}")
print(f"Signed: {pos_balance:+} | {neg_balance:-}")
```

## 4. Number Grouping with Underscores
Underscores can be used instead of commas (e.g., for code readability):

```python
population = 83200000
print(f"Comma: {population:,}")
print(f"Underscore: {population:_}")
```

## 5. Percentage with Precision
You can specify how many decimals to show in percentages:

```python
ratio = 0.72345
print(f"Rounded to 1 decimal: {ratio:.1%}")
print(f"Rounded to 2 decimals: {ratio:.2%}")
```
