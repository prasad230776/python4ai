## Exercise 1: Square the Numbers

You’re given a list of numbers.  
Write a `for` loop that prints the **square** of each number.

````python
numbers = [2, 4, 6, 8]

# Your code here
````

<details>
<summary>Solution - click to reveal</summary>

```
numbers = [2, 4, 6, 8]
for number in numbers:
    print(number ** 2)
```

## Exercise 2: Calculate Total Cost

You’re given a list of item prices.  
Write a `for` loop that adds up the total cost and prints the **final total**.


````python
item_prices = [5.99, 3.49, 12.99, 2.99]

# Your code here
````

<details>
<summary>Solution - click to reveal</summary>

```
item_prices = [5.99, 3.49, 12.99, 2.99]

total = 0
for price in item_prices:
    total += price

print("Total cost:", total)

```

</details>


## Exercise 3: Countdown Timer

Use a `for` loop with `range()` to count down from **5 to 1**,  
then print `"Blast off!"`.

````python
# Your code here
````

<details>
<summary>Solution - click to reveal</summary>

```
for number in range(5, 0, -1):
    print(number)

print("Blast off!")

```

</details>
