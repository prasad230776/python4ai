#### Exercise 1: Declarative String Conversion

**Task:** 
1. You have a list of lowercase names: `names = ["aarav", "dev", "amit"]`.
2. Convert this to a declarative approach using `map()` and a lambda function to capitalize each name (hint: use `name.capitalize()`).
3. Cast the result to a list and print it.

````python
# TODO: Convert lowercase list to capitalized list declaratively
````

<details>
<summary>💡 Click to see solution</summary>

```python
names = ["aarav", "dev", "amit"]
capitalized_names = list(map(lambda name: name.capitalize(), names))
print(capitalized_names) # ['Aarav', 'Dev', 'Amit']
```
</details>

#### Exercise 2: Discount Strategy Policies

**Task:** 
1. Write a function `calculate_final_price(price, strategy_func)`.
2. Define a discount policy function `flat_discount(price)` that subtracts Rs. 50 from the price.
3. Define a discount policy function `percent_discount(price)` that subtracts 20% from the price.
4. Test both discount policies on a starting price of Rs. 500.

````python
# TODO: Write strategy functions and process prices
````

<details>
<summary>💡 Click to see solution</summary>

```python
def calculate_final_price(price, strategy_func):
    return strategy_func(price)

def flat_discount(price):
    return price - 50

def percent_discount(price):
    return price * 0.8

price = 500
print("Flat strategy:", calculate_final_price(price, flat_discount))       # 450
print("Percent strategy:", calculate_final_price(price, percent_discount)) # 400.0
```
</details>

#### Exercise 3: Filter & Map Pipeline

**Task:** Given a list `numbers = [5, 12, 18, 3, 22, 9]`:
1. Use `filter()` to select only numbers greater than 10.
2. Use `map()` to multiply each selected number by 3.
3. Cast to a list and print the results.

````python
# TODO: Implement filter and map pipeline
````

<details>
<summary>💡 Click to see solution</summary>

```python
numbers = [5, 12, 18, 3, 22, 9]
filtered = filter(lambda x: x > 10, numbers)
mapped = map(lambda x: x * 3, filtered)
print(list(mapped)) # [36, 54, 66]
```
</details>

#### Exercise 4: Finding Max Value with Reduce

**Task:** Given a list `values = [4, 21, 13, 8, 33, 19]`:
1. Use `reduce()` from `functools` along with a lambda function to find the maximum value in the list (hint: the lambda compares `x` and `y` and returns the larger one).
2. Print the maximum value.

````python
# TODO: Use reduce to find maximum value
````

<details>
<summary>💡 Click to see solution</summary>

```python
from functools import reduce

values = [4, 21, 13, 8, 33, 19]
max_val = reduce(lambda x, y: x if x > y else y, values)
print(max_val) # 33
```
</details>
