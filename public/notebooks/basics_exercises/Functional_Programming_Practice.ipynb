### The Declarative Approach

#### Example: Doubling a List
Comparing step-by-step imperative loops against a declarative mapper pipeline.

```python
# Imperative Approach
numbers = [1, 2, 3, 4]
doubled_imp = []
for num in numbers:
    doubled_imp.append(num * 2)
print("Imperative:", doubled_imp)

# Declarative Approach
doubled_dec = list(map(lambda x: x * 2, numbers))
print("Declarative:", doubled_dec)
```

### Higher-Order Functions

#### Example: Passing a Function as Argument
Defining generic processors that accept dynamic execution callbacks.

```python
def process_numbers(numbers, operation):
    result = []
    for num in numbers:
        result.append(operation(num))
    return result

def square(x):
    return x * x

def cube(x):
    return x ** 3

nums = [1, 2, 3, 4, 5]
print("Squares:", process_numbers(nums, square))
print("Cubes:", process_numbers(nums, cube))
print("Add 10:", process_numbers(nums, lambda x: x + 10))
```

#### Example: Returning a Function
Creating customized operations dynamically using closures.

```python
def multiplier(factor):
    def multiply(number):
        return number * factor
    return multiply

double = multiplier(2)
triple = multiplier(3)

print(double(10))
print(triple(10))
```

#### Example: Strategy Pattern (Real-World Salary Policies)
Applying dynamic mathematical policies to salary inputs.

```python
def process_salary(salary, policy):
    return policy(salary)

def tax(salary):
    return salary * 0.9

def bonus(salary):
    return salary * 1.2

salary = 50000
print("Taxed salary:", process_salary(salary, tax))
print("Bonus salary:", process_salary(salary, bonus))
```

### Built-In Higher-Order Functions

#### Example: map() Function
Mapping item operations with iterators.

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))
```

#### Example: Calling a separate multi-line function inside map()
Defining a separate multi-line helper (such as grading marks) and passing it directly into map by referencing its function name.

```python
def get_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "F"

student_marks = [88, 45, 92, 67]
student_grades = list(map(get_grade, student_marks))
print(student_grades)
```

#### Example: filter() Function
Filtering values using conditional predicates.

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

#### Example: reduce() Function
Aggregating sequence values to a single result.

```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

### Chained Functional Pipelines

#### Example: Sum of Squares of Even Numbers
Chaining filter, map, and reduce actions.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

# Step-by-step pipeline
evens = filter(lambda x: x % 2 == 0, numbers)
squares = map(lambda x: x ** 2, evens)
total_sum = reduce(lambda x, y: x + y, squares)
print("Step-by-step sum:", total_sum)

# Single chained statement
chained_sum = reduce(lambda x, y: x + y, map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)))
print("Chained sum:", chained_sum)
```

#### Example: Pythonic List Comprehension Equivalent
Using generators/comprehensions alongside sum() for clean pythonic layouts.

```python
numbers = [1, 2, 3, 4, 5, 6]
total_sum = sum(x**2 for x in numbers if x % 2 == 0)
print(total_sum)
```
