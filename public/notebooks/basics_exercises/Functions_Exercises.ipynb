    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter

#### Exercise 4: Filtering passing scores

**Task:** Write a function `filter_passing_scores(*scores, min_passing=40)` that accepts any number of scores as positional arguments, and returns a list containing only the scores greater than or equal to `min_passing`.
*(Notice that `min_passing` is a keyword-only parameter because it comes after `*scores`!)*

Test it passing the scores `35, 78, 42, 90, 20` using default and custom `min_passing=50` values.

````python
# TODO: Define and test filter_passing_scores
````

<details>
<summary>💡 Click to see solution</summary>

```python
def filter_passing_scores(*scores, min_passing=40):
    passing = []
    for score in scores:
        if score >= min_passing:
            passing.append(score)
    return passing

print(filter_passing_scores(35, 78, 42, 90, 20)) # [78, 42, 90]
print(filter_passing_scores(35, 78, 42, 90, 20, min_passing=50)) # [78, 90]
```

#### Exercise 5: Modifying Global and Enclosing scopes

**Task:** 
1. Declare a global variable `login_attempts = 0`. Write a function `track_attempt()` that uses the `global` keyword to increment `login_attempts` by 1.
2. Write a function `parent_task()` containing an enclosing variable `status = "pending"` and a nested function `complete_task()`. Inside `complete_task()`, use the `nonlocal` keyword to change `status` to `"completed"`. Call `complete_task()` inside `parent_task()` and print the final value of `status` from `parent_task()`.

````python
# TODO: Perform global and nonlocal scope modifications
````

<details>
<summary>💡 Click to see solution</summary>

```python
# Part 1: Global scope
login_attempts = 0

def track_attempt():
    global login_attempts
    login_attempts += 1

track_attempt()
track_attempt()
print("Global Attempts:", login_attempts) # 2

# Part 2: Nonlocal scope
def parent_task():
    status = "pending"
    
    def complete_task():
        nonlocal status
        status = "completed"
        
    complete_task()
    print("Enclosing Status:", status)

parent_task() # Enclosing Status: completed
```

#### Exercise 6: Recursive Countdown

**Task:** Write a recursive function named `recursive_countdown(n)` that performs a countdown and prints numbers from `n` down to `1`.
- **Base Case:** If `n <= 0`, it prints `"Start! 🏁"` and returns.
- **Recursive Case:** It prints `n` and calls `recursive_countdown(n-1)`.

````python
# TODO: Write recursive_countdown function
````

<details>
<summary>💡 Click to see solution</summary>

```python
def recursive_countdown(n):
    if n <= 0:
        print("Start! 🏁")
        return
    print(n)
    recursive_countdown(n - 1)

recursive_countdown(3)
```
</details>
