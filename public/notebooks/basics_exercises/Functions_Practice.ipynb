### Problem 1
What will be the output of the following code?

```python
def greet(name):
    return "Hello, " + name + "!"

print(greet("Alice"))
print(greet("Bob"))
```

### Problem 2
Predict the output of this code:

```python
def add(a, b):
    return a + b

result = add(5, 10)
print(result)
print(add(result, 20))
```

### Problem 3
What will this code print?


```python
def compute(a, b):
    result = a * b
    if result > 10:
        return result - 3
    else:
        return result + 4

print(compute(3, 2))
print(compute(5, 3))
```

### Problem 4
This function uses a global variable. Can you predict the output?

```python
counter = 10

def increment(value):
    global counter
    counter += value
    return counter

print(increment(5))
print(increment(3))
print(counter)
```
