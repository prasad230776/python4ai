### Defining Functions

#### Example: Your First Function
Defining and calling a basic function.

```python
def greet():
    print("Hello, world!")
    print("Welcome to Python!")

# Call the function
greet()
```

#### Example: Calling Functions Multiple Times
Reusing functions repeatedly.

```python
def say_goodbye():
    print("Goodbye!")
    print("See you later!")

# Call it multiple times
say_goodbye()
say_goodbye()
say_goodbye()
```

#### Example: Functions with Logic
Using conditionals inside function blocks.

```python
def check_weather():
    temperature = 25
    if temperature > 30:
        print("It's hot!")
    else:
        print("Nice weather!")

# Use the function
check_weather()
```

#### Example: Variable Scope - Local Variables
Variables defined inside functions cannot be accessed outside.

```python
def calculate_price():
    price = 100
    tax = price * 0.1
    print(f"Total: {price + tax}")

calculate_price()

# Trying to print(price) here will raise NameError!
```

#### Example: Variable Scope - Global Variables
Reading variables defined at the module-level.

```python
discount_rate = 0.15  # Global variable

def apply_discount(price):
    discount = price * discount_rate  # Can read global variable
    return price - discount

result = apply_discount(100)
print(result)  # 85.0
```

#### Example: Modifying Global Variables
Declaring global scopes using `global` keyword.

```python
counter = 0  # Global variable

def increment():
    global counter  # Declare we want to modify the global variable
    counter += 1

increment()
increment()
print(counter)  # 2
```

#### Example: Best Practice - Parameters & Returns over Globals
Passing states cleanly rather than modifying globals.

```python
def add_amounts(current_total, amount):
    return current_total + amount

total = 0
total = add_amounts(total, 10)
total = add_amounts(total, 20)
print(total)  # 30
```

### Parameters & Arguments

#### Example: Inflexible vs Flexible Functions
Functions accepting parameters compared to static functions.

```python
def greet_alice():
    print("Hello, Amit!")

def greet(name):
    print(f"Hello, {name}!")

greet_alice()
greet("Aarav")
```

#### Example: Basic Multiple Parameters
Accepting name and age inputs.

```python
def introduce(name, age):
    print(f"My name is {name}")
    print(f"I am {age} years old")

introduce("Amit", 25)
```

#### Example: Positional Order Mapping
Verifying positional arguments sequence.

```python
def describe_pet(animal_type, pet_name):
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet("hamster", "Harry")
describe_pet("Harry", "hamster")
```

#### Example: Positional Calculations
Passing numeric variables sequentially.

```python
def calculate_total(price, tax_rate, discount):
    tax = price * tax_rate
    final_price = price + tax - discount
    print(f"Total: ${final_price}")

calculate_total(100, 0.08, 10)  # $98.0
```

#### Example: Default Values
Making parameters optional.

```python
def greet_default(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet_default("Amit")
greet_default("Aarav", "Hi")
```

#### Example: Keyword Arguments
Passing arguments explicitly by name.

```python
def create_profile(name, age, city):
    print(f"{name}, {age}, from {city}")

create_profile(city="NYC", age=25, name="Amit")
```

#### Example: Mixing Positional and Keyword Arguments
Enforcing correct calling order.

```python
def book_flight(destination, date, seat="window", class_type="economy"):
    print(f"Flight to {destination} on {date}. Seat: {seat}, Class: {class_type}")

book_flight("Paris", "Oct 15", class_type="business")
```

#### Example: Dangerous vs Safe Default Values
Understanding mutable default list side-effects.

```python
# Wrong mutable default
def add_item_bad(item, lst=[]):
    lst.append(item)
    return lst

# Right safe approach
def add_item_good(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item_bad("a"))
print(add_item_bad("b")) # Notice the unexpected list persistence!

print(add_item_good("a"))
print(add_item_good("b"))
```

### Return Values

#### Example: Printing vs Returning
Comparing display actions and value handovers.

```python
def add_print(a, b):
    print(a + b)

def add_return(a, b):
    return a + b

add_print(5, 3)
res = add_return(5, 3)
print("Result:", res)
```

#### Example: Storing Area Returns
Calculating and returning rectangle areas.

```python
def calculate_area(width, height):
    area = width * height
    return area

room_area = calculate_area(10, 12)
print(f"Room size: {room_area} sq ft")
```

#### Example: Expressions & Condition checks on returns
Using returned values inside math equations and logic statements.

```python
def double(number):
    return number * 2

total = double(5) + double(3)
print("Total:", total)

if double(7) > 10:
    print("Big number!")
```

#### Example: Returning Multiple Values (Tuples)
Returning values separated by commas.

```python
def get_min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = get_min_max([5, 2, 8, 1, 9])
print(f"Min: {minimum}, Max: {maximum}")
print(get_min_max([5, 2, 8, 1, 9])) # Tuple object
```

#### Example: String Display vs String Returns
Showing how returned strings can be processed further (e.g. upper()).

```python
def get_greeting_print(name):
    print(f"Hello, {name}!")

def get_greeting_return(name):
    return f"Hello, {name}!"

val_print = get_greeting_print("Amit")
print("Print version returns:", val_print)

val_return = get_greeting_return("Amit")
print("Return version processed:", val_return.upper())
```

#### Example: Explicit returns of None
Functions without explicit returns yielding None.

```python
def greet_none(name):
    print(f"Hello, {name}!")

result = greet_none("Amit")
print("Returned:", result)
```

### Flexible Argument Handling

#### Example: Starred Parameters (`*args`)
Collecting arbitrary positional arguments.

```python
def sum_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_all(1, 2))
print(sum_all(5, 10, 15))
```

#### Example: Starred Keyword Parameters (`**kwargs`)
Collecting arbitrary keyword arguments.

```python
def show_profile(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

show_profile(name="Amit", age=25, city="NYC")
```

#### Example: Order of Parameters Mixed
Declaring required, args, defaults, and kwargs parameters sequentially.

```python
def master_function(required_arg, *args, default_arg="val", **kwargs):
    print(f"Required: {required_arg}")
    print(f"args: {args}")
    print(f"Default: {default_arg}")
    print(f"kwargs: {kwargs}")

master_function("First", 1, 2, 3, default_arg="override", name="Aarav")
```

#### Example: Positional-Only Parameters (`/`)
Enforcing positional-only inputs.

```python
def print_name(first, last, /):
    print(f"{first} {last}")

print_name("Amit", "Smith")
# print_name(first="Amit", last="Smith") # Raises TypeError
```

#### Example: Keyword-Only Parameters (`*`)
Enforcing keyword-only inputs.

```python
def calculate_tax(price, *, rate):
    return price * rate

print(calculate_tax(100, rate=0.08))
# print(calculate_tax(100, 0.08)) # Raises TypeError
```

### Scopes & Nested Functions

#### Example: Nested Inner Functions
Declaring helper functions inside function scopes.

```python
def outer_function(message):
    def inner_function():
        print(f"Message from inner: {message}")
        
    inner_function()

outer_function("Hello from Outer!")
```

#### Example: LEGB Variables Lookup Sequence
Understanding local, enclosing, global, and built-in shadowing.

```python
x = "global"

def outer():
    x = "enclosing"
    def inner():
        x = "local"
        print(x) # prints local
    inner()

outer()
```

#### Example: Global Keyword Modifiers
Altering module global variables.

```python
counter = 0

def increment():
    global counter
    counter += 1

increment()
print(counter)
```

#### Example: Nonlocal Enclosing Variables Modifier
Altering enclosing variables using `nonlocal`.

```python
def outer_function():
    message = "Hello from Outer"
    
    def inner_function():
        nonlocal message
        message = "Hello from Inner!"
        
    inner_function()
    print(message)

outer_function()
```

### Recursion

#### Example: Recursive Countdown
Countdown with base cases.

```python
def countdown(n):
    if n <= 0:
        print("Blast off! 🚀")
        return
    print(n)
    countdown(n - 1)

countdown(3)
```

#### Example: Recursive Factorial Calculation
Computing factorials.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(4))
```

#### Example: Recursive Fibonacci sequence
Computing Fibonacci numbers.

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6))
```
