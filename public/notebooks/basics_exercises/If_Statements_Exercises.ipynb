## Exercise 1: Check Even or Odd

**Task:** Write an `if-else` statement to check if the number `num` is even or odd. Print `"Even"` or `"Odd"`.

````python
# TODO: Check if num is even or odd
num = 17
````

<details>
<summary>💡 Click to see solution</summary>

```python
num = 17
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```
</details>

## Exercise 4: Larger of Two Values

**Task:** Given variables `val1 = 45` and `val2 = 72`, write an `if-else` statement to find and print the larger value.

````python
# TODO: Find and print the larger of the two variables
val1 = 45
val2 = 72
````

## Exercise 5: Largest of Three Values

**Task:** Given variables `a = 56`, `b = 89`, and `c = 42`, write an `if-elif-else` statement to find and print the largest value.

````python
# TODO: Find and print the largest of three variables
a = 56
b = 89
c = 42
````

<details>
<summary>💡 Click to see solution</summary>

```python
a = 56
b = 89
c = 42

if a >= b and a >= c:
    print("Largest is:", a)
elif b >= a and b >= c:
    print("Largest is:", b)
else:
    print("Largest is:", c)
```
</details>

## Exercise 6: Largest of Four Values

**Task:** Given variables `w = 23`, `x = 78`, `y = 56`, and `z = 91`, write an `if-elif-else` statement to find and print the largest value.

````python
# TODO: Find and print the largest of four variables
w = 23
x = 78
y = 56
z = 91
````

<details>
