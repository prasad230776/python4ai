# Python If Statements - Follow-Along Practice

Use this notebook to practice using decision-making conditionals (`if`, `elif`, `else`, nested `if`, and logical combinations).

### Concept 1: Basic If Statement
Checking a condition and executing indented code if it evaluates to `True`.

**Objective of this code:** To check if a person's age meets the legal voting requirement (18 or older) and grant voting eligibility status.

```python
age = 18
if age >= 18:
    print("You can vote!")
    print("You're an adult.")
```

### Concept 2: If-Else Statements
Handling both the branch where the condition is `True` and the alternative branch where it is `False`.

**Objective of this code:** To evaluate the current weather temperature (checking if it is above 30 degrees) and print a customized comfort status response.

```python
temperature = 25
if temperature > 30:
    print("It's hot!")
else:
    print("Nice weather!")
```

### Concept 3: If-Elif-Else Chains
Evaluating multiple consecutive conditions in order.

**Objective of this code:** To automatically categorize a student's performance grade (A, B, C, or F) based on their numeric score range.

```python
score = 85
if score >= 90:
    print("A - Excellent!")
elif score >= 80:
    print("B - Good job!")
elif score >= 70:
    print("C - Keep it up!")
else:
    print("F - Need improvement")
```

### Concept 4: Combining Conditions
Using logical operators (`and`, `or`, `not`) to construct complex checks.

**Objective of this code:** To verify multiple credentials or contexts (e.g. if someone is old enough AND has a license to drive, or if it is a weekend OR holiday to check work status).

```python
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive!")

weekend = True
holiday = False
if weekend or holiday:
    print("No work today!")
```

### Concept 5: Nested If Statements
Placing an `if` block inside another `if` block.

**Objective of this code:** To perform multi-level checks for entering a venue (first checking if the visitor has a ticket, and then verifying their age group to determine supervisor requirements).

```python
has_ticket = True
visitor_age = 15

if has_ticket:
    if visitor_age >= 18:
        print("Enjoy the movie!")
    else:
        print("Need adult supervision.")
else:
    print("Buy a ticket first.")
```
