## Problem 2

You have 17 books and want to distribute them evenly among 4 shelves.

Calculate:
1. How many books each shelf gets if you divide them equally (use regular division `/`)
2. How many whole books each shelf gets (use integer division `//`)
3. How many books are left over (use the modulo operator `%`)

**Sample output:**
```
4.25
4
1
```

````python
# TODO: Calculate 17 / 4


# TODO: Calculate 17 // 4


# TODO: Calculate 17 % 4
````

### Solution

````python
# Regular division
print(17 / 4)

# Integer division
print(17 // 4)

# Modulo (remainder)
print(17 % 4)
````

## Problem 3

When you mix integers and floats in calculations, Python automatically handles the conversion.

For each expression below, predict whether the result will be an integer or a float. Then print both the result and its type to verify.

Expressions:
1. `5 + 3`
2. `5 + 3.0`
3. `15 // 3`
4. `15.0 // 3`

````python
# TODO: For each expression, print both the result and its type
# Hint: You can print multiple things at once like this: print(5 + 3, type(5 + 3))
````

### Solution

````python
print(5 + 3, type(5 + 3))
print(5 + 3.0, type(5 + 3.0))
print(15 // 3, type(15 // 3))
print(15.0 // 3, type(15.0 // 3))
````
