```python
from pydantic import BaseModel, ValidationError

class Book(BaseModel):
    title: str
    author: str
    pages: int
    price: float

# Test valid
try:
    book = Book(title="Python for AI", author="Dave", pages=320, price=29.99)
    print("Valid book:", book)
except ValidationError as e:
    print(e)

# Test invalid
try:
    bad_book = Book(title="Python for AI", author="Dave", pages="three hundred", price=None)
except ValidationError as e:
    print("\nValidation failed successfully:")
    print(e)
```
