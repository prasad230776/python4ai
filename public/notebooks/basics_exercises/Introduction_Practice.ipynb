## 1. The problem with Python (Dynamic Typing)
Python is dynamically typed, meaning variables can change type dynamically without errors. This can lead to unexpected issues in application code.

```python
age = 25
print("Age as int:", age, type(age))

age = "twenty-five" 
print("Age as str:", age, type(age))

age = ["25", 25, None]
print("Age as list:", age, type(age))
```

## 2. When things go wrong
Let's see what happens when a function expects specific types (like an integer for `age`) but receives invalid or malformed data at runtime.

```python
def create_user(data):
    user_id = data["id"]
    email = data["email"]
    age = data["age"]
    
    # Later in your code...
    birth_year = 2025 - age  # What if age is "unknown" or a string instead of an int?
    return birth_year

# Valid JSON-like data - works fine
try:
    print("Valid data birth year:", create_user({"id": 1, "email": "dave@example.com", "age": 25}))
except Exception as e:
    print("Error:", e)

# Invalid data - will crash the application
try:
    print(create_user({"id": 1, "email": None, "age": "unknown"}))
except Exception as e:
    print("Crash caught successfully:", type(e).__name__, ":", e)
```

## 3. Pydantic to the rescue
Pydantic validates data at runtime. You define what your data should look like, and Pydantic ensures it matches, raising clear error messages when validation fails.

```python
from pydantic import BaseModel, ValidationError

class User(BaseModel):
    id: int
    email: str
    age: int

# Valid data - works fine
try:
    user = User(id=1, email="dave@example.com", age=25)
    print("User instantiated successfully:", user)
except ValidationError as e:
    print(e)

# Invalid data - fails immediately with clear error
try:
    user = User(id=1, email=None, age="unknown")
except ValidationError as e:
    print("\nValidation failed as expected:")
    print(e)
```
