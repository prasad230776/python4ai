## Exercise 2: Popping Elements

**Task:** Given a list `tasks = ["Email client", "Write code", "Run tests", "Deploy project"]`:
1. Pop the last element and print it.
2. Pop the element at index 1 and print it.
3. Print the remaining task list.

````python
# TODO: Pop elements with and without index
tasks = ["Email client", "Write code", "Run tests", "Deploy project"]
````

<details>
<summary>💡 Click to see solution</summary>

```python
tasks = ["Email client", "Write code", "Run tests", "Deploy project"]
last_task = tasks.pop()      # Pop last ('Deploy project')
second_task = tasks.pop(1)   # Pop index 1 ('Write code')
print("Popped last:", last_task)
print("Popped second:", second_task)
print("Remaining:", tasks) # ['Email client', 'Run tests']
```
</details>

## Exercise 3: Sort vs Sorted

**Task:** Given a list `scores = [85, 70, 95, 60]`:
1. Print a new sorted list of scores *without* changing the original list.
2. Print the original `scores` list to verify it remains unchanged.
3. Sort the original `scores` list in-place and verify it has changed.

````python
# TODO: Sort list using sorted() and .sort()
scores = [85, 70, 95, 60]
````

<details>
<summary>💡 Click to see solution</summary>

```python
scores = [85, 70, 95, 60]
print("Sorted copy:", sorted(scores))   # [60, 70, 85, 95]
print("Original list:", scores)         # [85, 70, 95, 60]

scores.sort()
print("Original (in-place):", scores) # [60, 70, 85, 95]
```
</details>
