# Python Lists - Follow-Along Practice

Use this notebook to practice using Python lists: creating lists, accessing elements, slicing, modifying (append/insert/pop), sorting, and math functions.
Each concept is in a separate cell, followed by a blank code cell for you to practice.

### Concept 1: Creating and Accessing Lists
Define a list using square brackets `[]` and access items using zero-based indexing.

```python
fruits = ["apple", "banana", "cherry"]
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
```

### Concept 2: List Slicing
Accessing segments of lists using `list[start:stop]`.

```python
numbers = [10, 20, 30, 40, 50, 60]
print("First three:", numbers[0:3])
print("Last two:", numbers[-2:])
```

### Concept 3: Modifying Lists (append, insert, pop, pop with index)
Using `.append(val)` to add to the end, `.insert(idx, val)` to insert at an index, `.pop()` to remove from the end, and `.pop(idx)` to remove at an index.

```python
fruits = ["apple", "banana"]
fruits.append("cherry")           # Append "cherry" to the end
fruits.insert(1, "mango")         # Insert "mango" at index 1
print("After inserts:", fruits)

last_fruit = fruits.pop()         # Pop from the end
first_fruit = fruits.pop(0)       # Pop from index 0
print("Remaining:", fruits)
print("Popped last:", last_fruit)
print("Popped first:", first_fruit)
```

### Concept 4: Sorting Lists (sort vs sorted)
Comparing the in-place `.sort()` method and the built-in `sorted()` function.

```python
nums = [4, 1, 3, 2]

# 1. sorted() function creates a new sorted list copy
sorted_copy = sorted(nums)
print("Original after sorted():", nums)
print("New sorted copy:", sorted_copy)

# 2. .sort() method modifies the list in-place
nums.sort()
print("Original after .sort():", nums)
```

### Concept 5: Mathematical Operations (sum, min, max)
Using built-in mathematical functions on numeric lists.

```python
scores = [85, 90, 78, 92, 60]
print("Sum of scores:", sum(scores))
print("Minimum score:", min(scores))
print("Maximum score:", max(scores))
```
