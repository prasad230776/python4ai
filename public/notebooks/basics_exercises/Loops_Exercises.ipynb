````python
# TODO: Write the countdown using a while loop
countdown = 5
````

## Exercise 4: Find the Largest of Three Values Using a Loop

**Task:** Given a list of three numbers `numbers = [34, 87, 52]`, use a `for` loop to iterate through the list and find the largest number. Print the largest value.

**Hint:** Initialize a tracker variable `largest = numbers[0]` before starting the loop, then compare each number in the loop to `largest`.

````python
# TODO: Find the largest value in the list using a for loop
numbers = [34, 87, 52]
````

<details>
<summary>💡 Click to see solution</summary>

```python
numbers = [34, 87, 52]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number is:", largest)  # 87
```
</details>

## Exercise 5: Search with Break and Else

**Task:** Given `search_target = 7` and a list of elements `items = [3, 8, 2, 7, 5]`, write a `for` loop to find if `search_target` exists in the list. 
If found, print `"Found it!"` and exit the loop immediately using `break`. 
If the loop finishes without finding the value, print `"Not found in the list"` using a loop `else` block.

````python
# TODO: Write loop with break and else
items = [3, 8, 2, 7, 5]
search_target = 7
````

<details>
<summary>💡 Click to see solution</summary>

```python
items = [3, 8, 2, 7, 5]
search_target = 7

for item in items:
    if item == search_target:
        print("Found it!")
        break
else:
    print("Not found in the list")
```
</details>

## Exercise 6: Skip Odd Numbers

**Task:** Use a `for` loop to iterate from `1` to `10`. Write a condition inside the loop that checks if a number is odd, and if so, skips the rest of the loop block using `continue`. Print the remaining (even) numbers.

````python
# TODO: Use continue to skip odd numbers
````

<details>

```python

        continue
    print(num)
```

## Exercise 7: Breaking Nested Loops

**Task:** Write nested loops where the outer loop iterates `i` from `1` to `3` and the inner loop iterates `j` from `1` to `3`. 
Stop all iterations and break out of both loops immediately when the product `i * j` equals `4`. Use a boolean flag `stop_loops` to achieve this.

````python
# TODO: Write nested loops and exit outer loop when i * j == 4
````

<details>
<summary>💡 Click to see solution</summary>

```python
stop_loops = False

for i in range(1, 4):
    for j in range(1, 4):
        if i * j == 4:
            print(f"Breaking at i={i}, j={j}")
            stop_loops = True
            break
    if stop_loops:
        break
```
</details>

## Exercise 8: Decreasing Triangle (Nested Loops)

**Task:** Write nested loops to print a decreasing triangle of height `4` using **only** loop controls (no string multiplication).

**Hint:** The outer loop `i` ranges from `1` to `n` (where `n=4`). The inner loop `j` should count down from `n` to `i` with a negative step of `-1` (using `range(n, i - 1, -1)`), printing a star for each iteration without moving to a new line.

````python
# TODO: Generate the decreasing triangle using nested loops counting down from n to i
````

<details>
<summary>💡 Click to see solution</summary>

```python
n = 4
for i in range(1, n + 1):             # i ranges from 1 to n
    for j in range(n, i - 1, -1):     # j counts down from n to i
        print("*", end="")
    print()                           # New row
```
</details>
