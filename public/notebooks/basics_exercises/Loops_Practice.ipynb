# Python Loops - Follow-Along Practice

Use this notebook to practice using repeat structures (`for` loops with range variations, string loops, and `while` loops).
Each concept is in a separate cell, followed by a blank code cell for you to practice.

### Concept 1: Basic For Loop (Repeating N Times)
Using `range(stop)` to count starting at 0 up to stop-1.

```python
# Print numbers 0 through 4
for i in range(5):
    print(i)
```

### Concept 2: Range Start, Stop, and Step
Using `range(start, stop, step)` to control count bounds and step value.

```python
# Count from 1 to 5
for i in range(1, 6):
    print("1 to 5:", i)

# Count by 2s
for i in range(0, 10, 2):
    print("Even steps:", i)
```

### Concept 3: Looping Through a String
Iterating over each character of a string in sequence.

```python
name = "Python"
for letter in name:
    print(letter)
```

### Concept 4: Preview - Looping Through Lists
Iterating over collections of elements.

```python
colors = ["red", "blue", "green"]
for color in colors:
    print(f"I like {color}")
```

### Concept 5: While Loops
Repeating code as long as a boolean condition remains `True`.

```python
count = 0
while count < 5:
    print(f"Count is {count}")
    count = count + 1
```

### Concept 6: Loop Control - break & continue
Using `break` to exit loops early, and `continue` to skip iterations.

```python
# break example
print("break output:")
for num in range(1, 10):
    if num == 5:
        break
    print(num)

# continue example
print("\ncontinue output:")
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

### Concept 7: Loop Else
Using the `else` block with loops which triggers *only if* the loop completes without a `break`.

```python
numbers = [1, 2, 3, 4, 5]
search_for = 6

for num in numbers:
    if num == search_for:
        print("Found it!")
        break
else:
    print("Not found in the list.")
```

### Concept 8: Nested Loops & Breaking Outer Loops
Running loops inside loops and exiting all loops using a flag variable.

```python
# Nested loops multiplication table
print("Table preview:")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}x{j}={i*j}", end="  ")
    print()

# Breaking outer loop using a flag
print("\nBreaking outer loop:")
target = 5
found = False
for i in range(1, 4):
    for j in range(1, 4):
        if i + j == target:
            print(f"Target sum {target} found at {i} and {j}!")
            found = True
            break
    if found:
        break
```

### Concept 9: Star Patterns (Increasing & Decreasing Logic)
Building pattern shapes by understanding Increasing ranges, Decreasing ranges, and combining them.

```python
#increasing triangle pattern
n = 4

print("Increasing Triangle:")
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end="")
    print()
```

```python
#decreasing triangle pattern
print("\nDecreasing Triangle (counting down from n to i):")
for i in range(1, n + 1):
    for j in range(n, i - 1, -1):
        print("*", end="")
    print()
```

```python
#right aligned triangle pattern
print("\nRight-Aligned Triangle (Decreasing Spaces + Increasing Stars):")
for i in range(1, n + 1):
    # Decreasing spaces
    for j in range(i, n):
        print(" ", end="")
    # Increasing stars
    for k in range(1, i + 1):
        print("*", end="")
    print()
```
