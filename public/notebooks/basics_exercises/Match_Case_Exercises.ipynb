## Exercise 1: Command System

**Task:** Write a `match` statement for a string variable `command`. Match:
- `"start"` -> print `"System is launching..."`
- `"stop"` -> print `"System is shutting down..."`
- `"pause"` -> print `"System is paused."`
- Any other value -> print `"Error: Unknown command."`

````python
# TODO: Write the match statement
command = "pause"
````

<details>
<summary>💡 Click to see solution</summary>

```python
command = "pause"

match command:
    case "start":
        print("System is launching...")
    case "stop":
        print("System is shutting down...")
    case "pause":
        print("System is paused.")
    case _:
        print("Error: Unknown command.")
```
</details>

## Exercise 2: Seasons Matching

**Task:** Write a `match` statement for an integer variable `month` (1-12) to print the season:
- `12`, `1`, `2` -> print `"Winter"`
- `3`, `4`, `5` -> print `"Spring"`
- `6`, `7`, `8` -> print `"Summer"`
- `9`, `10`, `11` -> print `"Autumn"`
- Any other value -> print `"Invalid month number."`

````python
# TODO: Write the match statement combining seasons
month = 4
````

<details>
<summary>💡 Click to see solution</summary>

```python
month = 4

match month:
    case 12 | 1 | 2:
        print("Winter")
    case 3 | 4 | 5:
        print("Spring")
    case 6 | 7 | 8:
        print("Summer")
    case 9 | 10 | 11:
        print("Autumn")
    case _:
        print("Invalid month number.")
```
</details>

## Exercise 3: Age Category Guard

**Task:** Write a `match` statement for an integer `age` using conditional `if` guards:
- Under 13 -> print `"Child Category"`
- 13 to 19 inclusive -> print `"Teenager Category"`
- 20 and above -> print `"Adult Category"`

````python
# TODO: Write the match statement using if guards
age = 15
````

<details>
<summary>💡 Click to see solution</summary>

```python
age = 15

match age:
    case a if a < 13:
        print("Child Category")
    case a if 13 <= a <= 19:
        print("Teenager Category")
    case a if a >= 20:
        print("Adult Category")
```
