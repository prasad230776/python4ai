## 1. Basic Match Case

### Concept 1: Basic Match Case & Wildcard
Matching single exact values and using the `_` fallback case.

```python
status_code = 404

match status_code:
    case 200:
        print("Success - OK")
    case 404:
        print("Not Found")
    case 500:
        print("Internal Server Error")
    case _:
        print("Unknown Status Code")
```

## 2. Combining Patterns (OR)

### Concept 2: Match-Case with OR (`|`) Patterns
Matching multiple possible values in a single case block.

```python
day = "Sunday"

match day:
    case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":
        print("It's a weekday.")
    case "Saturday" | "Sunday":
        print("It's the weekend!")
    case _:
        print("Invalid day!")
```

## 3. Conditional Guards

### Concept 3: Match-Case with `if` Guards
Adding conditions to cases to evaluate expressions.

```python
number = 15

match number:
    case n if n < 0:
        print(f"{n} is negative")
    case 0:
        print("Zero")
    case n if n > 0:
        print(f"{n} is positive")
```
