#### Example: Creating a Module
Creating calculator functions inside a reusable python file.

```python
# calculator.py content mock
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

PI = 3.14159

print("Add:", add(5, 3))
print("PI constant:", PI)
```

#### Example: Importing a Module
Importing standard library modules like `math` and calling its functions.

```python
import math
result = math.sqrt(64)
print("Square root of 64:", result)
print("Math PI:", math.pi)
```

#### Example: Package imports and __all__ controls
Controlling exposed modules during package wildcard imports.

```python
# mypackage/__init__.py content mock
# Only expose 'utils' and keep 'database' private to the package
__all__ = ['utils']
print("Package initialized with __all__ whitelists:", __all__)
```

#### Example: How Python Finds Modules
Viewing directories Python searches in sequence using `sys.path`.

```python
import sys
print("Python search path locations:")
for path in sys.path[:3]: # Print first 3 directories
    print(path)
```

#### Example: The Special __name__ Variable
Checking execution mode (direct run vs import).

```python
def add(a, b):
    return a + b

if __name__ == "__main__":
    print("Testing calculator module...")
    print("Test result:", add(10, 5) == 15)  # True
```

#### Example: Inspecting Modules with dir()
Using the built-in `dir()` function to list definitions inside a module.

```python
import math
# See first 15 functions inside the math module
print(dir(math)[:15])
```

#### Example: Avoiding Wildcard Imports
Explicit imports are cleaner than using `*`.

```python
# Bad:
# from math import *

# Good:
from math import sqrt, pi
print("Square root of 25:", sqrt(25))
print("Pi:", pi)
```
