```python
from pydantic import BaseModel

class Address(BaseModel):
    city: str
    zipcode: str

class UserAccount(BaseModel):
    username: str
    address: Address

user = UserAccount(
    username="alice_dev",
    address={"city": "Mumbai", "zipcode": "400001"}
)
print(user)
```
