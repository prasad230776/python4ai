## Exercise 7: Roots of a Quadratic Equation

**Task:** Find the two roots of the quadratic equation $x^2 - 5x + 6 = 0$ (so coefficients are $a = 1$, $b = -5$, $c = 6$).
The quadratic formula is:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

**Hint:** Calculate the square root using exponentiation with `** 0.5`, and make sure to group the numerator and denominator using parentheses properly!

````python
# TODO: Calculate root1 and root2 for x^2 - 5x + 6 = 0
a = 1
b = -5
c = 6
````

<details>
<summary>💡 Click to see solution</summary>

```python
a = 1
b = -5
c = 6

# Calculate the square root of the discriminant
discriminant_sqrt = (b**2 - 4*a*c) ** 0.5

# Compute roots using parentheses
root1 = (-b + discriminant_sqrt) / (2 * a)
root2 = (-b - discriminant_sqrt) / (2 * a)

print("Root 1:", root1)
print("Root 2:", root2)
```
</details>
