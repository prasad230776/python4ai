### Concept 1: Arithmetic Operators
Basic calculations using `+`, `-`, `*`, `/`, `//`, `%`, and `**`.

```python
print(10 + 3)   # Addition
print(10 - 3)   # Subtraction
print(10 * 3)   # Multiplication
print(10 / 3)   # Division (always gives float)
print(10 // 3)  # Floor division (rounds down)
print(10 % 3)   # Modulo (remainder)
print(10 ** 3)  # Exponent (power)
```

### Concept 2: Order of Operations (PEMDAS)
Python follows the standard order of operations. Use parentheses to enforce order.

```python
result_no_parentheses = 2 + 3 * 4
result_parentheses = (2 + 3) * 4
print("Without parentheses:", result_no_parentheses)
print("With parentheses:", result_parentheses)
```

### Concept 3: Comparison Operators
Comparing values to produce booleans.

```python
age = 18
print(age == 18)    # Equal to
print(age != 21)    # Not equal to
print(age > 17)     # Greater than
print(age < 20)     # Less than
print(age >= 18)    # Greater than or equal
print(age <= 18)    # Less than or equal
```

### Concept 4: Logical Operators
Combining multiple boolean expressions using `and`, `or`, and `not`.

```python
age = 25
has_license = True

# AND - both must be true
can_drive = age >= 16 and has_license
print("Can drive?", can_drive)

# OR - at least one must be true
day = "Saturday"
is_weekend = day == "Saturday" or day == "Sunday"
print("Is weekend?", is_weekend)

# NOT - reverses the value
is_adult = age >= 18
is_child = not is_adult
print("Is child?", is_child)
```

### Concept 5: Truth Tables
Understanding logical evaluation combinations.

```python
print("AND:")
print(True and True)
print(True and False)
print(False and False)

print("\nOR:")
print(True or False)
print(False or False)

print("\nNOT:")
print(not True)
print(not False)
```

### Concept 6: Short-Circuit Evaluation (and / or)
Python evaluates expressions left-to-right and returns the value that determines the result.

```python
# Short-circuit with 'and' (returns first Falsy or last value)
print(2 and 3)       # 3
print(False and 3)   # False
print(0 and 3)       # 0
print(None and 3)    # None
print("" and 4)      # ""

# Short-circuit with 'or' (returns first Truthy or last value)
print(2 or 3)        # 2
print(False or 3)    # 3
print(0 or None)     # None
print("" or "default")# "default"
```

### Concept 7: Assignment Shortcuts
Updating a variable value in-place.

```python
score = 0
score += 10
print("Score + 10:", score)

x = 10
x += 5
x *= 2
print("Shortcut results:", x)
```

### Concept 8: Division Types Comparison
Regular division returns a float even if divided evenly.

```python
result_reg = 10 / 2
result_floor = 10 // 2
print("10 / 2 =", result_reg, type(result_reg))
print("10 // 2 =", result_floor, type(result_floor))
```

### Concept 9: Average Calculations & Parentheses
Calculating the average of numbers requires explicit grouping.

```python
average_bad = 10 + 20 + 30 / 3
average_good = (10 + 20 + 30) / 3
print("Incorrect average:", average_bad)
print("Correct average:", average_good)
```

### Concept 10: Assignment vs. Equality Comparison
Use `=` for assignments and `==` for comparisons.

```python
age = 18
if age == 18:
    print("Just turned adult!")
```
