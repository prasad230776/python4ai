## Exercise 2: Processing Student Marks using `*`

**Task:** Given a list of a student's marks `marks = [95, 88, 70, 60, 82]`:
1. Use starred unpacking to capture the first mark (as `first`), the last mark (as `last`), and pack all middle marks into a list named `middle`.
2. Print the value of `first`, `last`, and `middle`.

````python
# TODO: Unpack first, middle, and last marks
marks = [95, 88, 70, 60, 82]
````

<details>
<summary>💡 Click to see solution</summary>

```python
marks = [95, 88, 70, 60, 82]
first, *middle, last = marks
print("First mark:", first)   # 95
print("Middle marks:", middle) # [88, 70, 60]
print("Last mark:", last)     # 82
```
</details>

## Exercise 3: Merging Configurations

**Task:** Given two configuration dictionaries `default_profile = {"status": "active", "tier": "free"}` and `user_profile = {"tier": "premium", "location": "Bengaluru"}`:
1. Merge them into a single dictionary named `final_profile` using `**` unpacking.
2. Print the resulting dictionary (verify the user's `tier` overrides the default tier).

````python
# TODO: Merge dictionaries using double asterisks
default_profile = {"status": "active", "tier": "free"}
user_profile = {"tier": "premium", "location": "Bengaluru"}
````

<details>
<summary>💡 Click to see solution</summary>

```python
default_profile = {"status": "active", "tier": "free"}
user_profile = {"tier": "premium", "location": "Bengaluru"}

final_profile = {**default_profile, **user_profile}
print(final_profile)
# Output: {'status': 'active', 'tier': 'premium', 'location': 'Bengaluru'}
```
</details>
