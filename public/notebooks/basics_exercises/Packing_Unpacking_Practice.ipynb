# Python Packing and Unpacking - Follow-Along Practice

Use this notebook to practice tuple packing, basic unpacking, extended unpacking with `*`, dictionary merging with `**`, and loop unpacking.
Each concept is in a separate cell, followed by a blank code cell for you to practice.

### Concept 1: Tuple Packing
Assigning multiple values to a single variable without brackets packs them into a tuple.

```python
person = "Aarav", 25, "Engineer"
print("Packed object:", person)
print("Type:", type(person))
```

### Concept 2: Basic Unpacking
Extracting elements directly into separate variables. Variable count must match elements!

```python
employee = ("Amit", 30, "Developer")
name, age, role = employee
print("Name:", name)
print("Role:", role)
```

```python
# Demonstration of required counts (will raise ValueError)
try:
    x, y = employee
except ValueError as e:
    print("Error caught:", e)
```

### Concept 3: Starred Unpacking (`*`)
Using a single asterisk to pack the remaining elements into a list.

```python
numbers = [1, 2, 3, 4, 5]
first, *middle, last = numbers
print("First:", first)
print("Middle packed:", middle)
print("Last:", last)
```

### Concept 4: Dictionary Unpacking (`**`)
Using double asterisks to unpack and merge key-value pairs from multiple dictionaries.

```python
default_config = {"theme": "light", "port": 8000}
user_config = {"theme": "dark", "debug": True}

merged = {**default_config, **user_config}
print("Merged configurations:", merged)
```

### Concept 5: Loop Unpacking
Iterating over sequences of collections and unpacking values inside loop headers.

```python
# Loop on lists of tuples
items = [(1, "Apple"), (2, "Orange"), (3, "Banana")]
for idx, name in items:
    print(f"Item {idx} is {name}")
```

```python
# Loop on dictionary items()
prices = {"milk": 40, "bread": 25}
for product, price in prices.items():
    print(f"The price of {product} is Rs. {price}")
```
