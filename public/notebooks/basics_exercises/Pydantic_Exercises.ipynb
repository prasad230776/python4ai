### Exercise 5: Student Constraints Schema (Field Constraints)

**Task:** 
1. Define a Pydantic model `Student` with attributes:
   - `name: str` having minimum length 3.
   - `age: int` constrained between 5 and 100 inclusive (`ge=5, le=100`).
   - `grade: float` constrained between 0.0 and 10.0 inclusive.
2. Try to instantiate `Student` with invalid age or grade inside a try-except block and print the validation error.

````python
# TODO: Implement Student model and trigger validation error
````

<details>
<summary>💡 Click to see solution</summary>

```python
from pydantic import BaseModel, Field, ValidationError

class Student(BaseModel):
    name: str = Field(min_length=3)
    age: int = Field(ge=5, le=100)
    grade: float = Field(ge=0.0, le=10.0)

try:
    Student(name="Amit", age=3, grade=9.5)
except ValidationError as e:
    print(e)
```
</details>

### Exercise 6: Email Symbol Validation (Custom Validators)

**Task:** 
1. Create a BaseModel `EmailRegistration` containing a field `email: str`.
2. Add a `@field_validator` on `email` to verify that the value contains the symbol `@`. If not, raise a `ValueError` with a descriptive message.
3. Test it with an invalid email within a try-except block.

````python
# TODO: Define EmailRegistration model with custom validator and test it
````

<details>
<summary>💡 Click to see solution</summary>

```python
from pydantic import BaseModel, field_validator, ValidationError

class EmailRegistration(BaseModel):
    email: str
    
    @field_validator("email")
    @classmethod
    def must_have_at_symbol(cls, v: str) -> str:
        if "@" not in v:
            raise ValueError("Invalid email address (must contain @)")
        return v

try:
    EmailRegistration(email="invalid-email-string")
except ValidationError as e:
    print(e)
```
</details>

### Exercise 8: Application Config Settings (Pydantic Settings)

**Task:** 
1. Define a class `DatabaseSettings` inheriting from `BaseSettings`.
2. Set the configuration prefix to `db_`.
3. Add fields: `host: str = "localhost"`, `port: int = 5432`.
4. Set the environment variable `DB_HOST=127.0.0.1` and `DB_PORT=5433` using `os.environ`.
5. Instantiate `DatabaseSettings` and print the loaded properties.

````python
# TODO: Implement DatabaseSettings model and verify environment loading
````

<details>
<summary>💡 Click to see solution</summary>

```python
import os
from pydantic_settings import BaseSettings, SettingsConfigDict

os.environ["DB_HOST"] = "127.0.0.1"
os.environ["DB_PORT"] = "5433"

class DatabaseSettings(BaseSettings):
    model_config = SettingsConfigDict(env_prefix="db_")
    
    host: str = "localhost"
    port: int = 5432

db_settings = DatabaseSettings()
print(db_settings)
print("Host:", db_settings.host)
print("Port:", db_settings.port, type(db_settings.port))
```
