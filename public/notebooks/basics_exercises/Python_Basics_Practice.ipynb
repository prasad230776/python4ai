### Concept: Indentation Blocks
Python uses spaces to define nested blocks of code.

```python
temperature = 32
if temperature > 30:
    print("It's hot!")
    print("Turn on AC")
```

### Concept: Consistent and Nested Indentation
Always use consistent spaces (typically 4 spaces) per level.

```python
# Correct nested indentation
score = 95
if score > 90:
    print("A grade")
    if score == 100:
        print("Perfect!")

# Wrong indentation mixing spaces (uncomment to see IndentationError):
# if score > 90:
#   print("A grade")
#     if score == 100:
#       print("Perfect!")
```

## 4. Code Formatting

### Concept: Formatting Indentation
Using 4 spaces within functions and conditionals.

```python
def greet(name):
    if name:
        print(f"Hello, {name}!")
    else:
        print("Hello!")

greet("Amit")
greet("")
```

### Concept: Good vs. Bad Spacing
Spacing around math operators and in lists according to PEP 8.

```python
# Good spacing
x = 1 + 2
numbers = [1, 2, 3, 4]

# Bad spacing (harder to read, but works)
x=1+2
numbers=[1,2,3,4]

print(x, numbers)
```

### Concept: Line Length
Breaking up extremely long lines of code using parentheses.

```python
# Good - readable line length
long_string_good = (
    "This is a very long string that "
    "spans multiple lines for readability"
)

# Bad - too long
long_string_bad = "This is a very long string that spans multiple lines for readability but keeps going and going"

print(long_string_good)
print(long_string_bad)
```

## 5. Variables

### Concept: Creating First Variables
Assigning text, numbers, and boolean values.

```python
name = "Amit"
age = 25
is_student = True
print(name, age, is_student)
```

### Concept: Allowed Naming Conventions
Checking variable naming constraints.

```python
# Allowed names
user_name = "Dev"      # snake_case (standard)
userName = "Dev"       # camelCase (works but not Python style)
age2 = 30               # numbers at end are OK
_private = "secret"     # underscores at start are OK

# Not allowed names (uncomment to see SyntaxError):
# 2age = 30
# my-name = "Dev"
# my name = "Dev"
# class = "Python"
```

### Concept: snake_case Convention
PEP 8 standard requires snake_case (lowercase with underscores) for variable names.

```python
# Good Python style
first_name = "Amit"
user_age = 25
is_logged_in = True
shopping_cart_total = 49.99

# Avoid camelCase
firstName = "Amit"
userAge = 25
isLoggedIn = True
```

### Concept: Updating Variables
Re-assigning values or modifying current values.

```python
score = 0
print(score)  # Shows: 0

score = 10
print(score)  # Shows: 10

score = score + 5
print(score)  # Shows: 15
```

## 6. Comments

### Concept: Single-Line Comments
Using `#` to document your code.

```python
# This is a comment
print("Hello")  # This is also a comment

# You can have multiple lines
# of comments by starting
# each line with a hash

age = 25  # Store user's age
```

### Concept: Multi-Line Comments
Using triple quotes (`"""`) to write docstrings or block comments.

```python
"""
This is a multi-line comment.
It can span several lines.
Great for longer explanations.
"""

def calculate_tip(bill):
    """
    Calculate 20% tip for a restaurant bill.
    Takes the bill amount and returns the tip.
    """
    return bill * 0.20

print(calculate_tip(50))
```

### Concept: When to Use Comments
Explain *why* instead of *what* is happening.

```python
subtotal = 100.0

# Good: Explains why
# Using 1.0625 because sales tax in CA is 6.25%
total = subtotal * 1.0625

# Bad: States the obvious
# Multiply subtotal by 1.0625
total = subtotal * 1.0625
```

### Concept: Disabling Code Temporarily
Using comments to debug or preserve old references.

```python
print("Starting process...")
# print("Debug info")  # Uncomment for debugging
# old_method()  # Keeping for reference
```

### Concept: Outdated & Over-commenting Mistakes
Examples of poor comment accuracy or excess.

```python
bill = 50.0
# Bad (outdated): Says 15% but does 20%
# Calculate 15% tip
tip = bill * 0.20
    
# Good: Accurate
# Calculate 20% tip (standard)
tip = bill * 0.20

# Bad (over-commenting obvious parts):
x = 5  # Set x to 5
y = 10  # Set y to 10
    
# Good: Only explain complex parts
x = 5
y = 10
position = x + y  # Convert to linear index
```
