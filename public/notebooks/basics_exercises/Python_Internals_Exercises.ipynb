#### Exercise 1: Memory Address and Mutability

**Task:** 
1. Define a list `my_list = [10, 20]` and print its memory `id()`.
2. Modify it by calling `my_list.append(30)`.
3. Print `my_list` and its new `id()`. Verify if the memory address changed.

````python
# TODO: Modify list and check memory address IDs
````

<details>
<summary>💡 Click to see solution</summary>

```python
my_list = [10, 20]
old_id = id(my_list)
print("Old ID:", old_id)

my_list.append(30)
new_id = id(my_list)
print("New ID:", new_id)
print("Address changed?:", old_id != new_id) # False
```

#### Exercise 3: Identity vs Equality Comparison

**Task:** 
1. Create two lists `list_c = [4, 5]` and `list_d = [4, 5]`.
2. Print the results of comparing them with `==` and with `is`.
3. Explain why one is `True` and the other is `False`.

````python
# TODO: Compare lists and print identities vs equality
````

<details>
<summary>💡 Click to see solution</summary>

```python
list_c = [4, 5]
list_d = [4, 5]

print("list_c == list_d:", list_c == list_d) # True (values match)
print("list_c is list_d:", list_c is list_d) # False (different memory addresses)
```
</details>
