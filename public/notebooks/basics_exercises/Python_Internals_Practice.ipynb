### Mutability vs Immutability

#### Example: Mutable Objects (Lists)
Modifying a list in-place does not change its memory address ID.

```python
numbers = [1, 2, 3]
print(id(numbers))

numbers.append(4)
print(numbers)
print(id(numbers))
```

#### Example: Immutable Objects (Strings)
Modifying an immutable string creates a new string object at a different address.

```python
name = "Amit"
print(id(name))

name += " Smith"
print(id(name))
```

#### Example: Tuples with Mutable Elements
A tuple is immutable, but a mutable list inside it can still be modified.

```python
my_tuple = (1, [2, 3])
my_tuple[1].append(4)
print(my_tuple)
```

### Everything is an Object

#### Example: Functions as Objects
Assigning a function reference to a variable and checking its type.

```python
def greet():
    return "Hello!"

func_reference = greet
print(func_reference())
print(type(greet))
```

### Identity vs Equality

#### Example: == vs is comparison
Testing value equality vs identity references.

```python
list_a = [1, 2, 3]
list_b = [1, 2, 3]

print(list_a == list_b)
print(list_a is list_b)
print(id(list_a) == id(list_b))
```

#### Example: Integer Caching optimization
CPython caching of small integers.

```python
x = 100
y = 100
print(x is y)

a = 300
b = 300
print(a is b)
```

### Reference Counting

#### Example: Reference Counts
Checking referencing counts of objects.

```python
import sys

a = [1, 2, 3]
b = a
print(sys.getrefcount(a))
```
