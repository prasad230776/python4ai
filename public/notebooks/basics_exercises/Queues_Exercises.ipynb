## Exercise 1: Customer Support Line

**Task:** Write code using `collections.deque` representing a support desk line:
1. Enqueue three customers: `"Aarav"`, `"Amit"`, and `"Dev"`.
2. Dequeue (serve) the first two customers and print their names.
3. Print the name of the remaining customer left in the line.

````python
# TODO: Simulate customer support line using deque
````

<details>
<summary>💡 Click to see solution</summary>

```python
from collections import deque

support_line = deque()

# Enqueue
support_line.append("Aarav")
support_line.append("Amit")
support_line.append("Dev")

# Dequeue
c1 = support_line.popleft()
c2 = support_line.popleft()
print("Served customer 1:", c1)
print("Served customer 2:", c2)
print("Remaining in line:", support_line[0])
```

```python
from collections import deque

print_jobs = deque()
print_jobs.append("Invoice.pdf")
print_jobs.append("Report.docx")
print_jobs.append("Presentation.pptx")

while len(print_jobs) > 0:
    doc = print_jobs.popleft()
    print(f"Printing {doc}...")
```
