## 1. Using collections.deque (Recommended)

### Concept 1: Enqueue and Dequeue with `collections.deque`
Using `.append()` to add items to the back, and `.popleft()` to remove from the front efficiently ($O(1)$ time).

```python
from collections import deque

# Initialize queue
ticket_line = deque()

# Enqueue (add to back)
ticket_line.append("Aarav")
ticket_line.append("Amit")
ticket_line.append("Dev")
print("Initial line:", ticket_line)

# Dequeue (remove from front)
first_person = ticket_line.popleft()
print("Served:", first_person)
print("Remaining line:", ticket_line)
```

## 2. Lists as Queues (Inefficient)

### Concept 2: Standard Lists as Queues
Lists can use `.append()` and `.pop(0)`, but popping the first element takes $O(N)$ time because Python has to shift all remaining elements in memory.

```python
list_queue = ["Aarav", "Amit", "Dev"]
list_queue.append("Priya")

# Dequeue using pop(0)
served = list_queue.pop(0)  # Slow shifting operation
print("Served:", served)
print("Queue:", list_queue)
```
