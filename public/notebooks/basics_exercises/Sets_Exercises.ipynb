## Exercise 1: Removing Duplicates

**Task:** Given a list with duplicates `names_list = ["Aarav", "Amit", "Aarav", "Dev", "Amit"]`:
1. Convert it to a set to remove all duplicates.
2. Print the unique set.
3. Convert it back to a list named `unique_names` and print it.

````python
# TODO: Remove duplicates from names_list
names_list = ["Aarav", "Amit", "Aarav", "Dev", "Amit"]
````

<details>
<summary>💡 Click to see solution</summary>

```python
names_list = ["Aarav", "Amit", "Aarav", "Dev", "Amit"]
unique_set = set(names_list)
print("Unique Set:", unique_set) # {'Aarav', 'Amit', 'Dev'}

unique_names = list(unique_set)
print("Unique List:", unique_names) # ['Aarav', 'Amit', 'Dev'] (order may vary)
```

## Exercise 2: Set Operations (Union, Intersection, Sym-Diff)

**Task:** Given two skill sets: `skills_a = {"Python", "SQL", "Excel"}` and `skills_b = {"Python", "Java", "C++"}`.
1. Find and print the skills that appear in **both** sets (Intersection) using the `&` operator.
2. Find and print the skills unique to `skills_a` only (Difference) using the `-` operator.
3. Find the skills that are in A or B but **not both** (Symmetric Difference) using the `^` operator.

````python
# TODO: Perform intersection, difference, and symmetric difference operations
skills_a = {"Python", "SQL", "Excel"}
skills_b = {"Python", "Java", "C++"}
````

<details>
<summary>💡 Click to see solution</summary>

```python
skills_a = {"Python", "SQL", "Excel"}
skills_b = {"Python", "Java", "C++"}

print("Skills in both (Intersection):", skills_a & skills_b)         # {'Python'}
print("Skills unique to A (Difference):", skills_a - skills_b)        # {'SQL', 'Excel'}
print("Skills in either but not both (Sym-Diff):", skills_a ^ skills_b) # {'SQL', 'Excel', 'Java', 'C++'}
```
</details>
