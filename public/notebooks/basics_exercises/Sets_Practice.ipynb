# Python Sets - Follow-Along Practice

Use this notebook to practice using Python sets: storing unique elements, set algebra operations, and subsets.
Each concept is in a separate cell, followed by a blank code cell for you to practice.

### Concept 1: Sets Uniqueness and Modification
Defining sets using curly braces `{}` (without colons) and adding/removing elements.

```python
colors = {"red", "blue", "green", "red"}
print("Unique colors in set:", colors)  # Notice duplicate 'red' is gone

colors.add("yellow")
colors.remove("blue")
print("Modified set:", colors)
```

### Concept 2: Mathematical Set Operations
Performing unions (`|`), intersections (`&`), differences (`-`), and symmetric differences (`^`).

```python
set_a = {1, 2, 3}
set_b = {3, 4, 5}

print("Union (A | B):", set_a | set_b)
print("Intersection (A & B):", set_a & set_b)
print("Difference (A - B):", set_a - set_b)
print("Symmetric Difference (A ^ B):", set_a ^ set_b)
```

### Concept 3: Subsets & Supersets
Checking relationships using `.issubset()` (or `<=`) and `.issuperset()` (or `>=`).

```python
small = {1, 2}
big = {1, 2, 3, 4}

print("Is 'small' a subset of 'big'?:", small <= big)
print("Is 'big' a superset of 'small'?:", big >= small)
```
