## Exercise 1: String Formatting Options

**Task:** Given variables `user = "Aarav"` and `points = 1500`, write statements to output `"Aarav has score of 1500 points."`, using:
1. An f-string.
2. The `.format()` method with positional parameters.

````python
# TODO: Output the sentence using f-string and .format()
user = "Aarav"
points = 1500
````

<details>
<summary>💡 Click to see solution</summary>

```python
user = "Aarav"
points = 1500

# f-string
print(f"{user} has score of {points} points.")

# .format() positional
print("{} has score of {} points.".format(user, points))
```

## Exercise 3: Slicing with Mixed Boundaries & Steps

**Task:** Given the string `text = "MachineLearning"`, write slice expressions to:
1. Extract `"Machine"` using positive start and negative stop indices.
2. Extract `"Learning"` using negative start and positive stop indices.
3. Extract every 3rd character starting from index 1 to the end.
4. Reverse the string completely using negative steps.

````python
# TODO: Perform slicing operations on text
text = "MachineLearning"
````

<details>
<summary>💡 Click to see solution</summary>

```python
text = "MachineLearning"
print(text[0:-8])     # 1. "Machine"
print(text[-8:15])    # 2. "Learning"
print(text[1::3])     # 3. "aierg" (every 3rd character starting at index 1)
print(text[::-1])     # 4. "gninraeLenihcaM" (reversal)
```
</details>

## Exercise 5: Splitting and Joining Strings

**Task:** Split `items = "pen;pencil;paper;book"` into a list, replace the word `"pencil"` with `"marker"` in the list, and then join them back using a space-plus-space separator (` + `).

````python
# TODO: Split, edit list element, and join color string
items = "pen;pencil;paper;book"
````

<details>
<summary>💡 Click to see solution</summary>

```python
items = "pen;pencil;paper;book"
items_list = items.split(";")
items_list[1] = "marker"  # Replace pencil with marker
result = " + ".join(items_list)
print(result)  # "pen + marker + paper + book"
```
</details>
