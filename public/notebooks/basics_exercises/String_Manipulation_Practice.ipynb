## 1. String Operations & Formatting

### Concept 1: Concatenation and Repetition
Joining strings together using `+` and repeating them using `*`.

```python
first_name = "Priya"
last_name = "Sharma"

full_name = first_name + " " + last_name
print("Full name:", full_name)

separator = "-" * 20
print("Separator:", separator)
```

### Concept 2: String Formatting (f-strings vs .format())
Using modern f-strings and the older `.format()` method (positional and named).

```python
name = "Amit"
age = 25
role = "Developer"

# 1. f-strings (modern)
print(f"Hello, I am {name} and I am {age} years old.")

# 2. .format() positional placeholders
print("Hello, {} is a {}.".format(name, role))

# 3. .format() named placeholders
print("Hello, {user} is a {job}.".format(user=name, job=role))
```

## 2. String Indexing & Slicing

### Concept 3: Positive and Negative Indexing
Accessing single characters using indices (0 to len-1) or negative indices (-1 to -len).

```python
text = "Python"

# Positive indexing (left to right)
print(text[0])   # 'P'
print(text[2])   # 't'

# Negative indexing (right to left)
print(text[-1])  # 'n'
print(text[-3])  # 'h'
```

### Concept 4: Basic Slicing & Steps
Extracting substrings using `[start:stop]` or `[start:stop:step]`.

```python
text = "Python Programming"

print(text[0:6])   # "Python"
print(text[7:])    # "Programming"
print(text[0:6:2]) # "Pto" (every 2nd character)
```

### Concept 5: Mixed Indexing & Step Slicing
Slicing using combinations of positive/negative indices and positive/negative steps.

```python
text = "Python Programming"

# 1. Mixed positive & negative index boundaries
print(text[1:-2])    # Start positive index 1, stop before negative index -2
print(text[-11:15])  # Start negative index -11, stop before positive index 15

# 2. Mixed values in step direction
print(text[1:-1:2])  # Positive start/step, negative stop
print(text[-2:2:-3]) # Negative start/step, positive stop
```

### Concept 6: Slicing Edge Cases
Handling out-of-bounds indices and mismatched direction parameters.

```python
text = "Python"

# Out of bounds indices are handled safely
print("Out-of-bounds:", text[0:100])

# Mismatched directions return an empty string
print("Mismatched (positive step with backward range):", repr(text[5:2]))
print("Mismatched (negative step with forward range):", repr(text[2:5:-1]))
```

## 3. String Methods

### Concept 7: Case Changes and Cleaning
Using `.lower()`, `.upper()`, `.title()`, and `.strip()`.

```python
text = "  Python Programming  "
print(text.lower())
print(text.upper())
print(text.title())
print(repr(text.strip()))
```

### Concept 8: Finding and Replacing
Searching substring indices, checking starts/ends, and replacing keywords.

```python
message = "I love Python programming with Python"
print("Python" in message)
print(message.startswith("I"))
print(message.find("Python"))
print(message.replace("Python", "JavaScript"))
```

### Concept 9: Splitting and Joining
Using `.split()` and `.join()` for list-string conversions.

```python
data = "apple,banana,cherry"
fruits = data.split(",")
print("Fruits List:", fruits)

joined_data = " | ".join(fruits)
print("Joined:", joined_data)
```
