## Exercise 1: Coordinates and Modification

**Task:** 
1. Define a tuple named `office_location` containing the coordinates `(12.97, 77.59)`.
2. Try to change the second coordinate to `77.60` inside a `try-except` block to catch and print the resulting `TypeError`.
3. Correctly modify the coordinate by converting the tuple to a list first, and then converting it back to a tuple.

````python
# TODO: Define, attempt modification, and correctly update coordinates
````

<details>
<summary>💡 Click to see solution</summary>

```python
office_location = (12.97, 77.59)

try:
    office_location[1] = 77.60
except TypeError as e:
    print("Error caught:", e)

temp = list(office_location)
temp[1] = 77.60
office_location = tuple(temp)
print("Updated Office Location:", office_location) # (12.97, 77.6)
```

## Exercise 2: Membership and Sorting

**Task:** Given a tuple `grades = (85, 90, 78, 92)`:
1. Check if `90` exists in the tuple using `in` and print the boolean result.
2. Generate a sorted copy of the tuple (as a tuple) using `sorted()` and `tuple()`, then print the sorted tuple.

````python
# TODO: Check membership and sort tuple
grades = (85, 90, 78, 92)
````

<details>
<summary>💡 Click to see solution</summary>

```python
grades = (85, 90, 78, 92)
print("Is 90 present?:", 90 in grades) # True

sorted_grades = tuple(sorted(grades))
print("Sorted tuple:", sorted_grades) # (78, 85, 90, 92)
```
