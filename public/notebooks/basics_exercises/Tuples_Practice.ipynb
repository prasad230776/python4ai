# Python Tuples - Follow-Along Practice

Use this notebook to practice using Python tuples: creating immutable sequences, indexing/slicing, membership checks, and sorted copies.

### Concept 1: Creating and Accessing Tuples
Defining tuples using parentheses `()`. Remember single-item tuples need a trailing comma!

```python
coordinates = (12.97, 77.59)
single_item_tuple = (42,)
print("Type of single_item_tuple:", type(single_item_tuple))
print("Latitude:", coordinates[0])
```

### Concept 2: Tuple Slicing and Membership Checks
Using index range slices `tuple[start:stop]` and checking item existence using the `in` operator.

```python
colors = ("red", "green", "blue", "yellow")
print("Slice (1 to 3):", colors[1:3])

print("Is 'red' in colors?:", "red" in colors)
print("Is 'purple' in colors?:", "purple" in colors)
```

### Concept 3: Sorting Tuples
Since tuples are immutable, they don't have a `.sort()` method. We use `sorted()` instead, which returns a new sorted list.

```python
numbers = (5, 2, 8, 1)
sorted_list = sorted(numbers)
sorted_tuple = tuple(sorted_list)
print("Sorted list from tuple:", sorted_list)
print("Converted back to tuple:", sorted_tuple)
```
