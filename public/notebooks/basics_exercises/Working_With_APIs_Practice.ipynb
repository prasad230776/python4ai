### First API Call

#### Example: Basic API Query
Sending a GET request to the Open-Meteo API using coordinates and printing the returned dictionary.

```python
import requests

# Coordinates for Paris
latitude = 48.85
longitude = 2.35

url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current=temperature_2m"

response = requests.get(url)
data = response.json()
print("Raw JSON Response keys:", list(data.keys()))
print("Elevation:", data.get("elevation"))
```

### Extracting Data from JSON Responses

#### Example: Accessing Nested Fields
Extracting specific keys from dictionary trees returned by HTTP JSON payloads.

```python
temperature = data['current']['temperature_2m']
print(f"Temperature in Paris: {temperature}°C")
```

### Dynamic API Request Wrapper Functions

#### Example: Reusable Weather Function
Abstracting parameters into query functions to compare weather parameters across multiple cities.

```python
import requests

def get_weather(latitude, longitude):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current=temperature_2m,wind_speed_10m"
    response = requests.get(url)
    data = response.json()
    return data['current']['temperature_2m']

paris_temp = get_weather(48.85, 2.35)
london_temp = get_weather(51.50, -0.12)
tokyo_temp = get_weather(35.68, 139.69)

print(f"Paris Temperature: {paris_temp}°C")
print(f"London Temperature: {london_temp}°C")
print(f"Tokyo Temperature: {tokyo_temp}°C")
```
