#### Exercise 2: User Settings JSON serialization

**Task:** 
1. Create a Python dictionary representing user application settings (e.g., `{"theme": "dark", "notifications": True, "volume": 80}`).
2. Convert it to a JSON formatted string with an indent of 2, and print the string.
3. Load it back from the string into a python dictionary and print the value of the `theme` key.

````python
# TODO: Serialize and deserialize settings dict
````

<details>
<summary>💡 Click to see solution</summary>

```python
import json

settings = {"theme": "dark", "notifications": True, "volume": 80}
json_str = json.dumps(settings, indent=2)
print(json_str)

loaded_settings = json.loads(json_str)
print("Theme value:", loaded_settings["theme"]) # dark
```

#### Exercise 3: Store Inventory CSV Builder

**Task:** 
1. Define a list of dictionaries representing products: `[{"Item": "Book", "Price": 150, "Count": 25}, {"Item": "Pen", "Price": 10, "Count": 100}]`.
2. Write these items to a CSV file named `"store_inventory.csv"` using `csv.DictWriter`.
3. Be sure to write the header row first before writing rows.

````python
# TODO: Write products list to store_inventory.csv
````

<details>
<summary>💡 Click to see solution</summary>

```python
import csv

products = [
    {"Item": "Book", "Price": 150, "Count": 25},
    {"Item": "Pen", "Price": 10, "Count": 100}
]
headers = ["Item", "Price", "Count"]

with open("store_inventory.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader()
    writer.writerows(products)

# Output contents
with open("store_inventory.csv", "r") as f:
    print(f.read())
```
</details>

#### Exercise 4: Reading Marks CSV and Computing Average

**Task:** 
1. Write a CSV file named `"student_marks.csv"` with content:
   ```
   Name,Marks
   Aarav,85
   Amit,90
   Dev,75
   ```
2. Open and read it using `csv.DictReader`.
3. Parse each score to an integer and calculate the average score of the class.

````python
# TODO: Create student_marks.csv, read it, and compute average score
````

<details>
<summary>💡 Click to see solution</summary>

```python
import csv

with open("student_marks.csv", "w") as f:
    f.write("Name,Marks\n")
    f.write("Aarav,85\n")
    f.write("Amit,90\n")
    f.write("Dev,75\n")

total_marks = 0
count = 0
with open("student_marks.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        total_marks += int(row["Marks"])
        count += 1

print("Average score:", total_marks / count) # 83.33333333333333
```
