### Handling Text Data

#### Example: Writing Text Files
Writing lines to a raw text file using `open()` with mode `'w'`.

```python
with open("sample.txt", "w") as file:
    file.write("Hello, World!\n")
    file.write("Writing data is easy in Python.\n")
```

#### Example: Reading Text Files (All at once)
Reading the entire content of a file into a single string variable.

```python
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
```

#### Example: Reading Text Files (Line-by-Line)
Looping over a file object to read lines iteratively for memory efficiency.

```python
with open("sample.txt", "r") as file:
    for line in file:
        print(line.strip())
```

### Handling JSON Data

#### Example: JSON Serialization & Deserialization
Converting dictionaries to JSON strings (`dumps`) and parsing them back (`loads`).

```python
import json

user_profile = {
    "name": "Alice",
    "age": 25,
    "skills": ["Python", "Machine Learning"],
    "is_active": True
}

# 1. Convert Dict to JSON String
json_string = json.dumps(user_profile, indent=4)
print(json_string)

# 2. Convert JSON String back to Dict
data_dict = json.loads(json_string)
print("Name from Dict:", data_dict["name"])
```

#### Example: JSON File Operations
Writing JSON directly to a file (`dump`) and loading it back (`load`).

```python
import json

user_profile = {
    "name": "Alice",
    "age": 25,
    "skills": ["Python", "Machine Learning"],
    "is_active": True
}

# 3. Writing JSON to file
with open("profile.json", "w") as file:
    json.dump(user_profile, file, indent=4)

# 4. Reading JSON from file
with open("profile.json", "r") as file:
    loaded_data = json.load(file)
    print("Loaded Skills:", loaded_data["skills"])
```

### Handling CSV Data

#### Helper: Writing Setup Employee CSV File
Creating the mock CSV file for practice reading.

```python
with open("employees.csv", "w") as f:
    f.write("Name,Age,Role\n")
    f.write("Aarav,30,Developer\n")
    f.write("Amit,28,Data Analyst\n")
```

#### Example: Reading CSV as Lists
Reading rows step-by-step using standard `csv.reader`.

```python
import csv

with open("employees.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

#### Example: Reading CSV as Dictionaries
Mapping rows to header keys using `csv.DictReader`.

```python
import csv

with open("employees.csv", "r") as file:
    dict_reader = csv.DictReader(file)
    for row in dict_reader:
        print(f"Name: {row['Name']}, Role: {row['Role']}")
```

#### Example: Writing CSV Files using DictWriter
Writing headers and structured dictionaries using `csv.DictWriter`.

```python
import csv

fieldnames = ["Product", "Price", "Stock"]
products = [
    {"Product": "Laptop", "Price": 999.99, "Stock": 10},
    {"Product": "Mouse", "Price": 29.99, "Stock": 50}
]

with open("inventory.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    # Write header
    writer.writeheader()
    
    # Write data rows
    writer.writerows(products)

# Verify by printing contents
with open("inventory.csv", "r") as file:
    print(file.read())
```
