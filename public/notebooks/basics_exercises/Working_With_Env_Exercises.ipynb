3. Save these keys to a new file named `.env.example` where the value is replaced by empty space (e.g. `API_KEY=`).
4. Read and print the contents of `.env.example`.

````python
# TODO: Parse .env keys and write .env.example template
````

<details>
<summary>💡 Click to see solution</summary>

```python
with open(".env", "r") as infile, open(".env.example", "w") as outfile:
    for line in infile:
        if "=" in line:
            key, _ = line.split("=", 1)
            outfile.write(f"{key.strip()}=\n")

with open(".env.example", "r") as f:
    print(f.read())
```

3. Read the value of `APP_ENV` with a fallback default value of `"development"`.
4. Print the result.

````python
# TODO: Append APP_ENV, reload env variables, and print value
````

<details>
<summary>💡 Click to see solution</summary>

```python
import os
from dotenv import load_dotenv

with open(".env", "a") as f:
    f.write("\nAPP_ENV=production\n")

load_dotenv(override=True)
app_env = os.getenv("APP_ENV", "development")
print("App Environment:", app_env) # production
```

#### Exercise 3: Multi-Key Validation Checker

**Task:** 
1. Write a function `check_configuration(required_keys)` that accepts a list of string keys.
2. The function checks if each key is present in environment variables using `os.getenv()`.
3. If any key is missing, raise a `ValueError` specifying which key was missing.
4. Test the validation check function using a list like `["API_KEY", "SECRET_TOKEN"]` inside a try-except block.

````python
# TODO: Implement check_configuration and test it
````

<details>
<summary>💡 Click to see solution</summary>

```python
import os

def check_configuration(required_keys):
    for key in required_keys:
        if not os.getenv(key):
            raise ValueError(f"Missing required environment key: {key}")
    print("All required configurations found!")

try:
    check_configuration(["API_KEY", "SECRET_TOKEN"])
except ValueError as e:
    print("Error:", e)
```
