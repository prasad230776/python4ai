### Loading Environment Variables

#### Example: Reading variables via python-dotenv
Calling `load_dotenv` and reading variables into Python memory using `os.getenv`.

```python
import os
from dotenv import load_dotenv

# 1. Load variables from .env file
load_dotenv()

# 2. Access variables
api_key = os.getenv("API_KEY")
port = os.getenv("API_PORT")

print("API Key:", api_key)
print("Port:", port)
```

### Environment Variable Fallbacks

#### Example: Accessing Missing Keys with Fallback Defaults
Specifying fallback parameters inside `os.getenv()` in case variables are absent.

```python
import os

# Fallbacks to sqlite if DATABASE_URL is missing
database_url = os.getenv("DATABASE_URL", "sqlite:///default.db")
print("Database URL:", database_url)
```

### Enforcing Configuration Keys

#### Example: Graceful Validation Check
Raising errors programmatically to block application startup when critical secrets are missing.

```python
import os
secret_key = os.getenv("WEATHER_API_KEY")

if not secret_key:
    print("Configuration Error: Missing WEATHER_API_KEY environment variable.")
```

### Practical Example: Secure API Call

#### Example: Secure API Querying
Using load_dotenv and os.getenv to supply secret keys to external web endpoints dynamically.

```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("WEATHER_API_KEY")

# Check key first to prevent running errors on missing tokens
if not api_key:
    print("WEATHER_API_KEY not found in env. Skipping actual request to avoid runtime error.")
else:
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q=Paris"
    response = requests.get(url)
    print(response.json())
```
