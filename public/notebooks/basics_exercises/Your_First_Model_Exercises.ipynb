### Exercise 1: Book Dataclass Definition (Dataclasses)

**Task:** 
1. Define a dataclass `Book` with attributes: `title: str`, `author: str`, `pages: int`, and `price: float = 0.0` (with default value `0.0`).
2. Instantiate a `Book` and print the object to verify the generated `__repr__` output formatting.

````python
# TODO: Implement Book dataclass and instantiate it
````

<details>
<summary>💡 Click to see solution</summary>

```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    author: str
    pages: int
    price: float = 0.0

book = Book("Python Basics", "Aarav Sharma", 300, 450.00)
print(book)
```

### Exercise 3: Employee Schema (Pydantic BaseModel)

**Task:** 
1. Define a Pydantic schema model `Employee` with fields: `emp_id: int`, `name: str`, `department: str`, and `is_manager: bool = False`.
2. Instantiate `Employee` using a string number for `emp_id` (e.g. `"1001"`) and verify coercion succeeds.

````python
# TODO: Implement Employee model and test coercion
````

<details>
<summary>💡 Click to see solution</summary>

```python
from pydantic import BaseModel

class Employee(BaseModel):
    emp_id: int
    name: str
    department: str
    is_manager: bool = False

emp = Employee(emp_id="1001", name="Aarav", department="Engineering")
print(emp)
print(type(emp.emp_id))  # <class 'int'>
```
</details>
