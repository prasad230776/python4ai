## Exercise 3: Slicing and Indexing

### 🧠 Concept Refresher
- **Slicing Syntax**: `arr[start:stop:step]`.
- **Negative Slicing**: `arr[:-n]` cuts elements up to the last `n` values (e.g. `arr[:-2]` excludes the last two elements).
- **2D Slicing**: `matrix[row_slice, column_slice]`.
  - Specific row: `matrix[row_idx, :]`
  - Specific column: `matrix[:, col_idx]`
  - Sub-grid: `matrix[row_start:row_end, col_start:col_end]`
- **Boolean Indexing**: Filtering arrays using comparison operators (e.g. `arr[arr > 5]`). Returns a flat array of elements meeting the condition.

### 🧠 Concept Refresher
- **`reshape(rows, cols)`**: Rearranges dimensions. Size must match original.

## Exercise 7: Broadcasting

### 🧠 Concept Refresher
- **Broadcasting**: Automatically matches shapes of arrays of different shapes by virtually expanding smaller dimensions until shapes align (evaluating dimensions from right-to-left).
- **Scalar Broadcasting**: A single value expands to fit any array shape (e.g. `matrix + 10` adds 10 to every cell).
- **Vector-to-Matrix Broadcasting**: A vector matches matrix dimensions when dimensions match, e.g. a `(3,)` row vector broadcasts over a `(2,3)` matrix.
