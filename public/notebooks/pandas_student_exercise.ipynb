## Exercise 4: Updating & Transforming Columns

### 🧠 Concept Refresher
- **Adding Columns**: `df['new_col'] = df['col1'] + df['col2']`.

**Tasks:**
1. Create a new column named `total_activities` which is the sum of `study_hours` and `assignments_completed`.
2. Use `np.where()` to create a column `sleep_category` where if `sleep_hours` is 7 or more, it labels it `'Healthy'`, else `'Sleep Deprived'`.
3. Use `.apply()` with a `lambda` function to increment all values in `previous_score` by 2 marks (make sure to cap the score at 100, i.e., it cannot exceed 100).

````python
# Hint: df['new_col'] = df['col1'] + df['col2']

````

````python
# Hint: np.where(condition, value_if_true, value_if_false)

````

<summary>💡 Click to see solution</summary>

```python
# Task 1
df['total_activities'] = df['study_hours'] + df['assignments_completed']

# Task 2
df['sleep_category'] = np.where(df['sleep_hours'] >= 7, 'Healthy', 'Sleep Deprived')

# Task 3
df['previous_score'] = df['previous_score'].apply(lambda x: min(x + 2, 100))
```

## Exercise 5: Data Aggregation & Grouping

- **`df.groupby('group_column')[['col1', 'col2']].mean()`**: Groups rows by matching values in `'group_column'` and computes calculations (like `mean()`, `sum()`, `median()`) on selected columns within each sub-group.

2. Group by `sleep_category` and find the average `exam_score` and `attendance`.

````python
# Hint: df.groupby('group_col')[['col1', 'col2']].mean()

````

````python
# Hint: Group by 'sleep_category' and aggregate 'exam_score' and 'attendance'

````
