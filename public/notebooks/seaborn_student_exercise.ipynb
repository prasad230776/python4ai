## Exercise 5: Correlation Heatmaps

### 🧠 Concept Refresher
- **`df.select_dtypes(include=['number'])`**: Filters columns containing only numbers.
- **`df.corr()`**: Computes Pearson correlation coefficients.
- **`sns.heatmap(corr, annot=True)`**: Plots color grids representing correlations; `annot=True` prints labels/values inside the cells.

**Task:** Extract numerical columns from `df`, compute the correlation matrix, and plot a correlation heatmap using the `"coolwarm"` colormap (`cmap`) with cell annotations enabled.

````python
# Hint: 1. Select numeric types, 2. calculate correlation matrix, 3. plot heatmap

````

<details>
<summary>💡 Click to see solution</summary>

```python
numeric_df = df.select_dtypes(include=["number"])
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
</details>
